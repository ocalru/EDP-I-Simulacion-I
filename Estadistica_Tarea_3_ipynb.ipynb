El valor de **500** puntos representa el **puntaje m√≠nimo requerido** para que un estudiante extranjero sea considerado para la admisi√≥n seg√∫n el criterio del examen **TOEFL**. Como la distribuci√≥n de puntajes TOEFL se modela como una distribuci√≥n normal con media $ \mu = 510 $ y desviaci√≥n est√°ndar $ \sigma = 70 $, estandarizamos este puntaje para calcular la probabilidad de que un estudiante supere este umbral. $$
Z_{TOEFL} = \frac{500 - 510}{70} = -0.1429
$$

$$
P(TOEFL \geq 500) = P(Z \geq -0.1429) = 1 - P(Z < -0.1429) \approx 1 - 0.4432 = 0.5568
$$



El valor de **102** puntos representa el **puntaje m√≠nimo requerido** en la prueba de inteligencia (IQ) para que un estudiante extranjero sea considerado en el proceso de admisi√≥n. Estandarizamos este valor para determinar qu√© porcentaje de estudiantes alcanza o supera este umbral. ($ IQ ‚â• 102: $)

$$

    <font size="4" color="red"> Calculando la probabilidad de que un alumno pase la prueba TOEFL: </font>

$$
P(TOEFL \geq 500) = P(Z \geq -0.1429)
$$

Este valor representa la **probabilidad de que un estudiante saque al menos $ 500$ puntos en el examen TOEFL**, es decir, que **cumpla con el requisito m√≠nimo** para ser admitido.

Para calcular esta probabilidad, usamos la tabla de la distribuci√≥n normal est√°ndar, que da los valores acumulados **hasta cierto valor $Z$**, es decir, $ P(Z < z) $.

Entonces:

$$
P(Z \geq -0.1429) = 1 - P(Z < -0.1429)
$$

- $ P(Z < -0.1429) $ representa la **proporci√≥n de estudiantes que obtienen una puntuaci√≥n menor a 500** en el TOEFL, es decir, **los que no pasan el examen**.

- Al hacer $ 1 - P(Z < -0.1429) $, estamos calculando la **proporci√≥n complementaria**, es decir, **los estudiantes que s√≠ lo pasan**.

Aproximando con la tabla:

$$
P(Z < -0.1429) \approx 0.4432
$$

$$
P(Z \geq -0.1429) = 1 - 0.4432 = 0.5568
$$

Por lo tanto, **aproximadamente el $ 55.68 \% $ de los estudiantes extranjeros cumplen el requisito del TOEFL**.

    <font size="4" color="red"> a) Calculo de la probabiliad de que pasan exactamente una de las dos pruebas: </font>
</p>




Queremos calcular la probabilidad de que **un estudiante pase exactamente una de las dos pruebas**, es decir:

- Que **pase TOEFL pero no IQ**: $ A \cap \neg B $
- Que **pase IQ pero no TOEFL**: $ \neg A \cap B $

$$
P(\text{exactamente una}) = P(A \cap \neg B) + P(\neg A \cap B)
$$

Donde:

- $ A $: el estudiante **pasa TOEFL** (es decir, $ TOEFL ‚â• 500 $ )
- $ \neg A $: el estudiante **no pasa TOEFL**
- $ B $: el estudiante **pasa IQ** (es decir,$ IQ ‚â• 102 $)
- $ \neg B $: el estudiante **no pasa IQ**



Estamos sumando **dos casos mutuamente excluyentes**:

1. Que pase TOEFL pero no IQ
2. Que pase IQ pero no TOEFL

Estos dos eventos **no se pueden dar al mismo tiempo**, por lo tanto, la probabilidad total de ‚Äúpasar exactamente una‚Äù es la **suma de las probabilidades** de ambos escenarios.



Cuando decimos $ P(A \cap \neg B) $, nos referimos a la probabilidad de que **ocurran ambas cosas al mismo tiempo**:

- Que el estudiante **s√≠ pase TOEFL**
- Y que el estudiante **no pase IQ**

De forma similar, $ P(\neg A \cap B) $ representa que:

- **No pasa TOEFL**
- **S√≠ pasa IQ**



$$
P(A \cap \neg B) = P(A) \cdot P(\neg B)
$$
$$
P(\neg A \cap B) = P(\neg A) \cdot P(B)
$$

Esto es v√°lido **porque se nos dice que las variables TOEFL e IQ son independientes**. Es decir, **el resultado en un examen no afecta el resultado del otro**. Por eso podemos usar la regla:

$$
P(X \cap Y) = P(X) \cdot P(Y) \quad \text{si X e Y son independientes}
$$



Sustituyendo valores:

$$
P(A) = 0.5568, \quad P(B) = 0.7475
$$
$$
P(\neg B) = 1 - P(B) = 0.2525, \quad P(\neg A) = 1 - P(A) = 0.4432
$$

$$
P(\text{exactamente una}) = 0.5568 \cdot 0.2525 + 0.4432 \cdot 0.7475
$$
$$
= 0.1406 + 0.3315 = \boxed{0.4721 \text{ (47.21%)}}
$$

Esto significa que **aproximadamente el $ 47.21 \% $ de los estudiantes pasan exactamente una de las dos pruebas**.

 <p align="center">
    <font size="4" color="red"> c) Pasan al menos una de las dos pruebas </font>
</p>




Queremos calcular la probabilidad de que **un estudiante pase al menos una** de las dos pruebas (TOEFL o IQ). En notaci√≥n de conjuntos, esto se expresa como:

$$
P(A \cup B)
$$

Donde:
- El estudiante **pasa TOEFL** (evento $ A $)
- El estudiante **pasa IQ** (evento $ B $)
- El estudiante pasa **ambas** pruebas (evento $ A \cap B $)

Es decir, estamos contando **todos los casos posibles** en los que el estudiante **pasa al menos una** de las dos pruebas.

$$
P(A \cup B) = P(A) + P(B) - P(A \cap B)
$$

Cuando se suma directamente $ P(A) + P(B) $, estamos **contando dos veces** a los estudiantes que pasan ambas pruebas (porque est√°n en $ A $ **y** en $ B $). Para **evitar ese doble conteo**, restamos una vez la intersecci√≥n:

- $ P(A) $: probabilidad de pasar TOEFL
- $ P(B) $: probabilidad de pasar IQ
- $ P(A \cap B) $: probabilidad de pasar ambas pruebas

Entonces:

$$
P(\text{al menos una}) = P(\text{TOEFL}) + P(\text{IQ}) - P(\text{ambas})
$$

Sustituyendo los valores:

$$
P(A) = 0.5568, \quad P(B) = 0.7475, \quad P(A \cap B) = P(A) \cdot P(B) = 0.5568 \cdot 0.7475 \approx 0.4162
$$

$$
P(A \cup B) = 0.5568 + 0.7475 - 0.4162 = 0.8881
$$
