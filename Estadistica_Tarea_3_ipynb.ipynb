Estandarizando las variables:

$$
Z = \frac{X - \mu}{\sigma}
$$

El valor de **500** puntos representa el **puntaje m√≠nimo requerido** para que un estudiante extranjero sea considerado para la admisi√≥n seg√∫n el criterio del examen **TOEFL**. Como la distribuci√≥n de puntajes TOEFL se modela como una distribuci√≥n normal con media $ \mu = 510 $ y desviaci√≥n est√°ndar $ \sigma = 70 $, estandarizamos este puntaje para calcular la probabilidad de que un estudiante supere este umbral. ( $TOEFL ‚â• 500$):

$$
Z_{TOEFL} = \frac{500 - 510}{70} = -0.1429
$$

$$
P(TOEFL \geq 500) = P(Z \geq -0.1429) = 1 - P(Z < -0.1429) \approx 1 - 0.4432 = 0.5568
$$



El valor de **102** puntos representa el **puntaje m√≠nimo requerido** en la prueba de inteligencia (IQ) para que un estudiante extranjero sea considerado en el proceso de admisi√≥n. La distribuci√≥n de resultados en esta prueba se modela como una normal con media $ \mu = 112 $ y desviaci√≥n est√°ndar $ \sigma = 15 $. Estandarizamos este valor para determinar qu√© porcentaje de estudiantes alcanza o supera este umbral. ($ IQ ‚â• 102: $)

$$
Z_{IQ} = \frac{102 - 112}{15} = -0.6667
$$

$$
P(TOEFL \geq 500) = P(Z \geq -0.1429)
$$

Este valor representa la **probabilidad de que un estudiante saque al menos $ 500$ puntos en el examen TOEFL**, es decir, que **cumpla con el requisito m√≠nimo** para ser admitido.

Para calcular esta probabilidad, usamos la tabla de la distribuci√≥n normal est√°ndar, que da los valores acumulados **hasta cierto valor $Z$**, es decir, $ P(Z < z) $.

Entonces:

$$
P(Z \geq -0.1429) = 1 - P(Z < -0.1429)
$$

- $ P(Z < -0.1429) $ representa la **proporci√≥n de estudiantes que obtienen una puntuaci√≥n menor a 500** en el TOEFL, es decir, **los que no pasan el examen**.

- Al hacer $ 1 - P(Z < -0.1429) $, estamos calculando la **proporci√≥n complementaria**, es decir, **los estudiantes que s√≠ lo pasan**.

Aproximando con la tabla:

$$
P(Z < -0.1429) \approx 0.4432
$$

$$
P(Z \geq -0.1429) = 1 - 0.4432 = 0.5568
$$

Por lo tanto, **aproximadamente el $ 55.68 \% $ de los estudiantes extranjeros cumplen el requisito del TOEFL**.

$$
P(IQ \geq 102) = P(Z \geq -0.6667)
$$

Este valor representa la **probabilidad de que un estudiante obtenga al menos $ 102 $ puntos en la prueba de IQ**, es decir, que **cumpla con el puntaje m√≠nimo requerido** para ser considerado en el proceso de admisi√≥n.

Para encontrar esta probabilidad, usamos la distribuci√≥n normal est√°ndar, que nos da la probabilidad acumulada de que un valor Z est√© **por debajo** de cierto umbral.

Entonces:

$$
P(Z \geq -0.6667) = 1 - P(Z < -0.6667)
$$

- $ P(Z < -0.6667) $ representa la **proporci√≥n de estudiantes que tienen un IQ menor a 102**, es decir, **los que no alcanzan el puntaje requerido**.

- Al hacer $ 1 - P(Z < -0.6667) $, obtenemos la **probabilidad de que un estudiante tenga un IQ de 102 o m√°s**, es decir, **los que s√≠ pasan el criterio de admisi√≥n por IQ**.

## Paso 2: Probabilidades

### a) Pasan exactamente una de las dos pruebas

\[
P(\text{exactamente una}) = P(A \cap \neg B) + P(\neg A \cap B)
\]

\[
= P(A)(1 - P(B)) + (1 - P(A))P(B)
\]

\[
= 0.5568 \cdot 0.2525 + 0.4432 \cdot 0.7475 \approx 0.1406 + 0.3315 = \boxed{0.4721 \text{ (47.21%)}}
\]

---

### b) Pasan ambas pruebas

\[
P(\text{ambas}) = P(A \cap B) = P(A) \cdot P(B) = 0.5568 \cdot 0.7475 = \boxed{0.4162 \text{ (41.62%)}}
\]

---

### c) Pasan al menos una de las dos pruebas

\[
P(A \cup B) = P(A) + P(B) - P(A \cap B)
\]

\[
= 0.5568 + 0.7475 - 0.4162 = \boxed{0.8881 \text{ (88.81%)}}
\]

---

### d) No pasan ninguna de las dos pruebas

\[
P(\text{ninguna}) = 1 - P(A \cup B) = 1 - 0.8881 = \boxed{0.1119 \text{ (11.19%)}}
\]
