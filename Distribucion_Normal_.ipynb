<p align="center">
    <font size="10" color="lightblue">Demostracion que la media de una distribucion normal estandar es 0. </font>
</p>



La función de densidad de una **distribución normal** con media $ \mu $ y desviación estándar $ \sigma $ es:

$$
f(x) = \frac{1}{\sigma \sqrt{2\pi}}\, e^{ -\frac{1}{2} \left( \frac{x - \mu}{\sigma} \right)^2 }, \quad x \in \mathbb{R}
$$



Si tomamos $ \mu = 0 $ y $ \sigma = 1 $, la función de densidad se convierte en:

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-x^2/2}
$$

Esta es la **función de densidad de la distribución normal estándar**, es decir, una normal con media 0 y varianza 1.



La media (esperanza matemática) de una variable aleatoria continua con densidad $ f(x) $ se define como:

$$
\mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x)\, dx
$$

Esta es una **integral impropia de tipo infinito**, ya que el dominio de integración es infinito. Formalmente, se define como:

$$
\int_{-\infty}^{\infty} x f(x)\, dx = \lim_{a \to \infty} \int_{-a}^{a} x f(x)\, dx
$$

Sustituimos la función de densidad de la normal estándar:

$$
\mathbb{E}[X] = \lim_{a \to \infty} \int_{-a}^{a} x \cdot \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}\, dx
$$

Sacamos la constante fuera de la integral:

$$
\mathbb{E}[X] = \frac{1}{\sqrt{2\pi}} \lim_{a \to \infty} \int_{-a}^{a} x e^{-\frac{x^2}{2}}\, dx
$$

La función que aparece en la integral es:

$$
g(x) = x e^{-x^2/2}
$$

Esta función es **impar**, ya que:

$$
g(-x) = (-x) e^{-(-x)^2/2} = -x e^{-x^2/2} = -g(x)
$$

Cuando una función es impar y se integra sobre un intervalo simétrico respecto al origen, se cumple:

$$
\int_{-a}^{a} g(x)\, dx = 0
$$

Por lo tanto, al tomar el límite:

$$
\lim_{a \to \infty} \int_{-a}^{a} g(x)\, dx = 0
$$

Entonces:

$$
\mathbb{E}[X] = \frac{1}{\sqrt{2\pi}} \cdot 0 = 0
$$





En general, para una distribución normal $ \mathcal{N}(\mu, \sigma^2) $, se cumple:

$$
\mathbb{E}[X] = \mu
$$

Por lo tanto, si $ \mu = 0 $, entonces:

$$
\mathbb{E}[X] = \mu = 0
$$

Es decir, **la media y la esperanza matemática coinciden** en una distribución normal, y en este caso ambas son 0.




<p align="center">
    <font size="10" color="lightblue">¿Que es f(x)?</font>
</p>

La función de densidad $ f(x) $ es la **derivada de la función de distribución acumulada** $ F(x) $, la cual nos da la probabilidad de que la variable aleatoria sea menor o igual a cierto valor:

$$
F(x) = \mathbb{P}(X \leq x)
$$

Entonces, la densidad se define como:

$$
f(x) = \frac{d}{dx} F(x)
$$

Esto quiere decir que $ f(x) $ **mide la velocidad con la que se acumula la probabilidad** a medida que avanzamos por el eje real. En otras palabras, $ f(x) $ nos da la **tasa de cambio** de la acumulación de probabilidad ( $ F(x) $ ) a lo largo de los valores de $ x $.

- Si $ f(x) $ es grande, significa que la probabilidad se está acumulando rápidamente alrededor de $ x $.
- Si $ f(x) $ es pequeña, significa que la acumulación de probabilidad es más lenta.
- Si $ f(x) = 0 $, no hay acumulación de probabilidad en ese punto.

Por ejemplo, para una distribución normal:

- La mayor parte de la probabilidad se concentra cerca de la media $ \mu $, por eso allí $ f(x) $ es máximo.
- En las colas de la distribución (valores muy alejados de $ \mu $),  $ f(x) $ es pequeña: la probabilidad se acumula más lentamente.



<p align="center">
    <font size="10" color="lightblue">Codigo para visualizar que E[X] = μ = 0 en una distribucion normal estandar. </font>
</p>

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la distribución normal

mu = 0       # esperanza matemática (media poblacional) para un distribucion normal estandar
sigma = 1    # desviación estándar para una distribucion normal estandar
n = 10000000     # tamaño de la muestra

# Generar una muestra aleatoria
muestra = np.random.normal(mu, sigma, n) # se usa la funcion np.random.normal(loc, scale, size), la cual genera una muestra aleatorio de tamaño "size"
# la cual sigue una distribucion normal.

#se definen los parametros para este caso:

# loc = mu (La media de la distribución.)
# scale = sigma (La desviación estándar.)
# size = n 	(El número de valores aleatorios que se van a generar)





# Calcular la media muestral
media_muestral = np.mean(muestra) # se usa la funcion "mean()" para calcular la media de la muestra generada.



print(f"Esperanza matemática (valor esperado): {mu}") # se mostrara el valor de mu
print(f"Media muestral (calculada con {n} datos): {media_muestral}") # se mostrara el valor de "media_muestral"
print(f"Error absoluto: {abs(media_muestral - mu)}") # se calculara y mostrara el valor "abs(media_muestral - mu)"", usando la funcion "abs()"
# que es la funcion valor absoluto. Esto es para calcular el error absoluto de la aproximacion del parametro mu

# Graficar la distribución
plt.hist(muestra, bins=1000, density=True, alpha=0.6, color='skyblue')
plt.axvline(mu, color='red', linestyle='dashed', linewidth=2, label='Esperanza matemática')
plt.axvline(media_muestral, color='green', linestyle='dotted', linewidth=2, label='Media muestral')
plt.title('Comparación entre la media muestral y la esperanza matematica para una distribucion normal estandar.')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()
```
