La funci√≥n de densidad de la distribuci√≥n normal est√°ndar es:

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}, \quad x \in \mathbb{R}
$$

La media (esperanza matem√°tica) de una variable aleatoria continua con densidad $ f(x) $ se define como:

$$
\mathbb{E}[X] = \int_{-\infty}^{\infty} x f(x)\, dx
$$


Esta es una **integral impropia de tipo infinito**, ya que el dominio de integraci√≥n es infinito. Formalmente, se define como:

$$
\int_{-\infty}^{\infty} x f(x)\, dx = \lim_{a \to \infty} \int_{-a}^{a} x f(x)\, dx
$$

Es decir, evaluamos la integral como el **l√≠mite** de una integral definida sobre un intervalo sim√©trico finito que se expande hacia el infinito.


$$
\mathbb{E}[X] = \int_{-\infty}^{\infty} x \cdot \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}\, dx
$$

Sacamos la constante fuera de la integral:

$$
\mathbb{E}[X] = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} x e^{-\frac{x^2}{2}}\, dx
$$


La funci√≥n que aparece en la integral es:

$$
g(x) = x e^{-x^2/2}
$$

Esta funci√≥n es **impar**, ya que:

$$
g(-x) = (-x) e^{-(-x)^2/2} = -x e^{-x^2/2} = -g(x)
$$

Cuando una funci√≥n es impar y se integra sobre un intervalo sim√©trico respecto al origen, se cumple:

$$
\int_{-a}^{a} g(x)\, dx = 0 \quad \Rightarrow \quad \lim_{a \to \infty} \int_{-a}^{a} g(x)\, dx = 0
$$

Entonces:

$$
\mathbb{E}[X] = \frac{1}{\sqrt{2\pi}} \cdot 0 = 0
$$

La media de la distribuci√≥n normal est√°ndar es **0**, porque la integral impropia de una funci√≥n impar sobre $ (-\infty, \infty) $ es 0.


La varianza de una variable aleatoria continua se define como:

$$
\text{Var}(X) = \mathbb{E}[X^2] - (\mathbb{E}[X])^2
$$

Ya demostramos que $ \mathbb{E}[X] = 0 $, as√≠ que:

$$
\text{Var}(X) = \mathbb{E}[X^2]
$$



Partimos de la definici√≥n de esperanza:

$$
\mathbb{E}[X^2] = \int_{-\infty}^{\infty} x^2 f(x)\, dx
$$

Donde:

$$
f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
$$

Sustituimos:

$$
\mathbb{E}[X^2] = \int_{-\infty}^{\infty} x^2 \cdot \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}\, dx
$$

Sacamos la constante:

$$
\mathbb{E}[X^2] = \frac{1}{\sqrt{2\pi}} \int_{-\infty}^{\infty} x^2 e^{-\frac{x^2}{2}}\, dx
$$



Este es un resultado cl√°sico conocido:

$$
\int_{-\infty}^{\infty} x^2 e^{-ax^2} dx = \frac{\sqrt{\pi}}{2a^{3/2}}, \quad \text{para } a > 0
$$

En nuestro caso, \( a = \frac{1}{2} \), as√≠ que:

$$
\int_{-\infty}^{\infty} x^2 e^{-\frac{x^2}{2}} dx = \frac{\sqrt{\pi}}{2 \cdot \left(\frac{1}{2}\right)^{3/2}} = \frac{\sqrt{\pi}}{2} \cdot \left(\frac{2^{3/2}}{1}\right) = \sqrt{\pi} \cdot \sqrt{2}
$$

Por tanto:

$$
\mathbb{E}[X^2] = \frac{1}{\sqrt{2\pi}} \cdot \sqrt{\pi} \cdot \sqrt{2} = 1
$$

#### ‚úÖ Conclusi√≥n

La varianza de la distribuci√≥n normal est√°ndar es:

$$
\text{Var}(X) = \mathbb{E}[X^2] = 1
$$


### üìò Definici√≥n formal de la funci√≥n de densidad de la distribuci√≥n normal

La **distribuci√≥n normal** (o gaussiana) con media \( \mu \in \mathbb{R} \) y varianza \( \sigma^2 > 0 \), denotada por:

$$
X \sim \mathcal{N}(\mu, \sigma^2)
$$

tiene una **funci√≥n de densidad de probabilidad** dada por:

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(x - \mu)^2}{2\sigma^2} \right), \quad x \in \mathbb{R}
$$

---

#### Propiedades importantes:

- Es una funci√≥n continua y suave (infinitamente derivable).
- Su dominio es todo \( \mathbb{R} \).
- El valor esperado (media) es \( \mathbb{E}[X] = \mu \).
- La varianza es \( \text{Var}(X) = \sigma^2 \).
- Es sim√©trica respecto a \( x = \mu \).
- El √°rea bajo la curva es 1:

$$
\int_{-\infty}^{\infty} f(x) \, dx = 1
$$

$$
