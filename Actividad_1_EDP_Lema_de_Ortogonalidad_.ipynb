Se demostrara si la funcion

$$\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right)\,dx $$

es igual a 0.

La función $ \cos\left(\frac{n\pi x}{L} \right) $ es par:

$$
\cos(-x) = \cos(x)
$$

Por lo tanto:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right)\,dx = 2 \int_0^L \cos\left(\frac{n\pi x}{L}\right)\,dx
$$


Sea:

$$
u = \frac{n\pi x}{L} \Rightarrow du = \frac{n\pi}{L} dx \Rightarrow dx = \frac{L}{n\pi} du
$$

Cuando $ x = 0 \Rightarrow u = 0 $

Cuando $ x = L \Rightarrow u = n\pi $

Entonces:

$$
\int_0^L \cos\left(\frac{n\pi x}{L}\right)\,dx = \int_0^{n\pi} \cos(u) \cdot \frac{L}{n\pi}\,du
= \frac{L}{n\pi} \int_0^{n\pi} \cos(u)\,du
$$



$$
\int_0^{n\pi} \cos(u)\,du = \sin(u)\big|_0^{n\pi} = \sin(n\pi) - \sin(0) = 0 - 0 = 0
$$

Por lo tanto:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right)\,dx = 2 \cdot \frac{L}{n\pi} \cdot 0 = 0
$$



$$
\int_{-L}^{L} \sin\left(\frac{n\pi x}{L}\right)\,dx = 0
$$

Se demostrara si la integral:



$$
\int_{-L}^{L} \sin\left(\frac{n\pi x}{L}\right)\,dx
$$

Es igual a $ 0 $


La función:

$$
f(x) = \sin\left(\frac{n\pi x}{L}\right)
$$

es una función **impar**, porque:

$$
f(-x) = \sin\left(-\frac{n\pi x}{L}\right) = -\sin\left(\frac{n\pi x}{L}\right) = -f(x)
$$

Para funciones impares $ f(x) $ integradas en intervalos simétricos $ [-a, a] $:

$$
\int_{-a}^{a} f(x)\,dx = 0
$$

Entonces:

$$
\int_{-L}^{L} \sin\left(\frac{n\pi x}{L}\right)\,dx = 0 \quad \text{(por simetría)}
$$

Se quiere demostrar que:

$$
\int_{-L}^{L} \cos\left(\frac{n \pi x}{L}\right) \cos\left(\frac{m \pi x}{L}\right) \, dx =
\begin{cases}
0, & \text{si } n \neq m, \\
L, & \text{si } n = m \neq 0,
\end{cases}
$$

Donde $ n, m \in \mathbb{Z}^+ $ (enteros positivos).


Se tiene que la integral:
$$
\int_{-L}^L \cos\left(\frac{n \pi x}{L}\right) \cos\left(\frac{m \pi x}{L}\right) dx
$$

Usando la **identidad trigonométrica**:

$$
\cos A \cos B = \frac{1}{2} \left[ \cos(A - B) + \cos(A + B) \right].
$$

Aplicamos esto con:

$$
A = \frac{n \pi x}{L}, \quad B = \frac{m \pi x}{L}.
$$

Entonces:

$$
\cos\left(\frac{n \pi x}{L}\right) \cos\left(\frac{m \pi x}{L}\right) = \frac{1}{2} \left[ \cos\left(\frac{(n - m)\pi x}{L}\right) + \cos\left(\frac{(n + m)\pi x}{L}\right) \right].
$$

La integral queda:

$$
\int_{-L}^L \cos\left(\frac{n \pi x}{L}\right) \cos\left(\frac{m \pi x}{L}\right) dx =
\frac{1}{2} \int_{-L}^L \cos\left(\frac{(n - m)\pi x}{L}\right) dx +
\frac{1}{2} \int_{-L}^L \cos\left(\frac{(n + m)\pi x}{L}\right) dx.
$$



Resolviendo la primera integral:

$$
I_1 = \frac{1}{2} \int_{-L}^{L} \cos\left(\frac{(n - m)\pi x}{L}\right) dx
$$

Encontrando la primitiva:

$$
\int \cos(ax)\, dx = \frac{1}{a} \sin(ax) + C
$$

Donde $ a = \frac{(n - m)\pi}{L} $

$$
\int \cos\left(\frac{(n - m)\pi x}{L}\right) dx = \frac{L}{(n - m)\pi} \sin\left(\frac{(n - m)\pi x}{L}\right)
$$

Evaluando entre $-L$ y $L$:**

$$
I_1 = \frac{1}{2} \cdot \left[ \frac{L}{(n - m)\pi} \sin\left(\frac{(n - m)\pi x}{L} \right) \right]_{-L}^{L}
$$

$$
= \frac{1}{2} \cdot \frac{L}{(n - m)\pi} \left[ \sin((n - m)\pi) - \sin(-(n - m)\pi) \right]
$$

$$
= \frac{1}{2} \cdot \frac{L}{(n - m)\pi} \cdot \left[ \sin((n - m)\pi) + \sin((n - m)\pi) \right] = \frac{L}{(n - m)\pi} \cdot \sin((n - m)\pi)
$$

$$
= \frac{L}{(n - m)\pi} \cdot 0 = 0 \quad \text{(porque } \sin(k\pi) = 0 \text{ para todo } k \in \mathbb{Z})
$$

Evaluando la segunda integral:

$$
I_2 = \frac{1}{2} \int_{-L}^{L} \cos\left(\frac{(n + m)\pi x}{L}\right) dx
$$

Calculando la primitiva:

$$
\int \cos\left(\frac{(n + m)\pi x}{L}\right) dx = \frac{L}{(n + m)\pi} \sin\left(\frac{(n + m)\pi x}{L} \right)
$$

Evaluando entre $-L$ y $ L $:**

$$
I_2 = \frac{1}{2} \cdot \left[ \frac{L}{(n + m)\pi} \sin\left(\frac{(n + m)\pi x}{L} \right) \right]_{-L}^{L}
$$

$$
= \frac{1}{2} \cdot \frac{L}{(n + m)\pi} \left[ \sin((n + m)\pi) - \sin(-(n + m)\pi) \right]
$$

$$
= \frac{1}{2} \cdot \frac{L}{(n + m)\pi} \left[ \sin((n + m)\pi) + \sin((n + m)\pi) \right]
= \frac{L}{(n + m)\pi} \cdot \sin((n + m)\pi)
$$

$$
= \frac{L}{(n + m)\pi} \cdot 0 = 0 \quad \text{(porque } \sin(k\pi) = 0 \text{ para todo } k \in \mathbb{Z})
$$
Se tiene que:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right) \cos\left(\frac{m\pi x}{L}\right) dx = 0 \quad \text{para } n \ne m
$$


Caso especial: $ n = m $


Partimos de la integral original:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right) \cos\left(\frac{m\pi x}{L}\right) dx
$$

En el caso especial en que $ n = m $, esta expresión se convierte en:

$$
\int_{-L}^{L} \cos^2\left(\frac{n\pi x}{L}\right) dx
$$

Esto ocurre porque estamos multiplicando una función consigo misma. Es decir, al ser $ n = m $, ambas funciones son idénticas, y la integral se convierte en la del cuadrado de una sola función:

$$
\cos\left(\frac{n\pi x}{L}\right) \cdot \cos\left(\frac{n\pi x}{L}\right) = \cos^2\left(\frac{n\pi x}{L}\right)
$$

Por lo tanto, la integral queda:

$$
\int_{-L}^{L} \cos^2\left(\frac{n\pi x}{L}\right) dx
$$




Usamos la identidad trigonométrica:

$$
\cos^2(\theta) = \frac{1 + \cos(2\theta)}{2}
$$

Entonces:

$$
\int_{-L}^{L} \cos^2\left(\frac{n\pi x}{L}\right) dx = \int_{-L}^{L} \frac{1 + \cos\left(\frac{2n\pi x}{L}\right)}{2} dx
$$

$$
= \frac{1}{2} \int_{-L}^{L} 1 \, dx + \frac{1}{2} \int_{-L}^{L} \cos\left(\frac{2n\pi x}{L}\right) dx
$$



**Primera parte:**

$$
\frac{1}{2} \int_{-L}^{L} 1 \, dx = \frac{1}{2} \cdot [x]_{-L}^{L} = \frac{1}{2} (L - (-L)) = \frac{1}{2} \cdot 2L = L
$$

**Segunda parte**

$$
\frac{1}{2} \int_{-L}^{L} \cos\left(\frac{2n\pi x}{L}\right) dx
$$

La primitiva es:

$$
\int \cos(ax) \, dx = \frac{1}{a} \sin(ax)
\quad \text{con } a = \frac{2n\pi}{L}
$$

Entonces:

$$
\int \cos\left(\frac{2n\pi x}{L}\right) dx = \frac{L}{2n\pi} \sin\left(\frac{2n\pi x}{L}\right)
$$

Evaluando:

$$
\left[ \frac{L}{2n\pi} \sin\left(\frac{2n\pi x}{L}\right) \right]_{-L}^{L}
= \frac{L}{2n\pi} \left[ \sin(2n\pi) - \sin(-2n\pi) \right] = \frac{L}{2n\pi} (0 + 0) = 0
$$

Entonces se tiene que:

$$
\int_{-L}^{L} \cos^2\left(\frac{n\pi x}{L}\right) dx = L
$$

Queremos demostrar que:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right)\cos\left(\frac{m\pi x}{L}\right)\, dx =
\begin{cases}
0, & \text{si } n \neq m \\
L, & \text{si } n = m \neq 0
\end{cases}
$$


Usamos la identidad trigonométrica:

$$
\cos A \cos B = \frac{1}{2}\left[\cos(A - B) + \cos(A + B)\right]
$$

Aplicamos esto con:

- $ A = \frac{n\pi x}{L} $
- $ B = \frac{m\pi x}{L} $

Entonces:

$$
\cos\left(\frac{n\pi x}{L}\right)\cos\left(\frac{m\pi x}{L}\right) =
\frac{1}{2}\left[
\cos\left(\frac{(n - m)\pi x}{L}\right) + \cos\left(\frac{(n + m)\pi x}{L}\right)
\right]
$$

Sustituyendo la integral:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right)\cos\left(\frac{m\pi x}{L}\right)\,dx =
\frac{1}{2} \int_{-L}^{L} \left[
\cos\left(\frac{(n - m)\pi x}{L}\right) + \cos\left(\frac{(n + m)\pi x}{L}\right)
\right] dx
$$



 Caso 1: $ n \neq m $

En este caso, $ n - m \neq 0$ y $n + m \neq 0$, por lo tanto:

con $ k = n - m $ o $ k = n + m $, y en ambos casos $ k \neq 0 $ cuando $ n \neq m $.

Como $ n $ y  $ m $ son números enteros, entonces $ k $ también es un número entero distinto de cero cuando $ n \neq m $.

Entonces, sustituyendo $ k $ en ambas integrales:

$$
\int_{-L}^{L} \cos\left(\frac{k\pi x}{L}\right) dx = 0 \quad \text{cuando } k \neq 0
$$

Entonces:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right)\cos\left(\frac{m\pi x}{L}\right)\,dx = 0
$$

Resultado: **0 si $ n \ne m $**


Caso 2: $ n = m \neq 0 $

La integral se convierte en:

$$
\int_{-L}^{L} \cos^2\left(\frac{n\pi x}{L}\right) dx
$$

Usamos la identidad:

$$
\cos^2 A = \frac{1}{2}(1 + \cos(2A))
$$

Entonces:

$$
\cos^2\left(\frac{n\pi x}{L}\right) = \frac{1}{2}\left(1 + \cos\left(\frac{2n\pi x}{L}\right)\right)
$$

Y la integral queda:

$$
\int_{-L}^{L} \cos^2\left(\frac{n\pi x}{L}\right) dx = \frac{1}{2} \int_{-L}^{L} \left(1 + \cos\left(\frac{2n\pi x}{L}\right) \right) dx
$$

Evaluamos ambas partes:

- $ \int_{-L}^{L} 1 \, dx = 2L$
- $\int_{-L}^{L} \cos\left(\frac{2n\pi x}{L}\right) dx = 0$

Por lo tanto:

$$
\int_{-L}^{L} \cos^2\left(\frac{n\pi x}{L}\right) dx = \frac{1}{2}(2L) = L
$$

</p>

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right) \sin\left(\frac{m\pi x}{L}\right) dx
$$

Usando la identidad trigonométrica

$$
\cos A \sin B = \frac{1}{2} \left[ \sin(A + B) - \sin(A - B) \right]
$$

con

$$
A = \frac{n\pi x}{L}, \quad B = \frac{m\pi x}{L}
$$

Entonces:

$$
\cos\left(\frac{n\pi x}{L}\right) \sin\left(\frac{m\pi x}{L}\right) = \frac{1}{2} \left[
\sin\left(\frac{(n + m)\pi x}{L}\right) - \sin\left(\frac{(n - m)\pi x}{L}\right)
\right]
$$

Definiendo  $ k_1 $ y $ k_2 $

Se tiene:

$$
k_1 = n + m, \quad k_2 = n - m
$$

Dado que $ n $ y $ m $ son enteros, entonces $ k_1 $ y $ k_2 $ también son enteros.



Sustituyendo en la integral:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right) \sin\left(\frac{m\pi x}{L}\right) dx = \frac{1}{2} \int_{-L}^{L} \left[
\sin\left(\frac{k_1 \pi x}{L}\right) - \sin\left(\frac{k_2 \pi x}{L}\right)
\right] dx
$$



Sabemos que la función seno es impar, es decir:

$$
\sin(-x) = -\sin(x)
$$

Por lo tanto, para cualquier entero $ k \neq 0 $, en particular para $ k_1 $ y $ k_2 $:

$$
\int_{-L}^{L} \sin\left(\frac{k_1 \pi x}{L}\right) dx = 0, \quad \int_{-L}^{L} \sin\left(\frac{k_2 \pi x}{L}\right) dx = 0
$$

porque la integral de una función impar en un intervalo simétrico $[-L, L]$ siempre es cero.




<p align="center">
    <font size="10" color="lightblue"> Ejercicio 2: </font>
</p>

Partimos de la **serie de Fourier** de una función $ f(x) $ en el intervalo $ [-L, L] $:

$$
f(x) = \frac{1}{2} a_0 + \sum_{n=1}^{\infty} \left( a_n \cos\left( \frac{n\pi x}{L} \right) + b_n \sin\left( \frac{n\pi x}{L} \right) \right)
$$

Queremos demostrar que:

$$
a_0 = \frac{1}{L} \int_{-L}^{L} f(x) \, dx
$$
$$
a_n = \frac{1}{L} \int_{-L}^{L} f(x) \cos\left( \frac{n\pi x}{L} \right) dx
$$
$$
b_n = \frac{1}{L} \int_{-L}^{L} f(x) \sin\left( \frac{n\pi x}{L} \right) dx
$$



Lema de Ortogonalidad Trigonométrica:

Para $ m, n \in \mathbb{N} $, se cumplen las siguientes integrales:

- $$
\int_{-L}^{L} \cos\left( \frac{n\pi x}{L} \right) \cos\left( \frac{m\pi x}{L} \right) dx =
\begin{cases}
0 & \text{si } n \ne m \\
L & \text{si } n = m
\end{cases}
$$

- $$
\int_{-L}^{L} \sin\left( \frac{n\pi x}{L} \right) \sin\left( \frac{m\pi x}{L} \right) dx =
\begin{cases}
0 & \text{si } n \ne m \\
L & \text{si } n = m
\end{cases}
$$

- $$
\int_{-L}^{L} \cos\left( \frac{n\pi x}{L} \right) \sin\left( \frac{m\pi x}{L} \right) dx = 0 \quad \text{para todo } n, m
$$

Cálculo de $ a_0 $:

Partinedo de la expresión de la serie de Fourier:

$$
f(x) = \frac{1}{2} a_0 + \sum_{n=1}^{\infty} \left( a_n \cos\left( \frac{n\pi x}{L} \right) + b_n \sin\left( \frac{n\pi x}{L} \right) \right)
$$

Integramos ambos lados en el intervalo $ [-L, L] $:

$$
\int_{-L}^{L} f(x) \, dx = \int_{-L}^{L} \left[ \frac{1}{2} a_0 + \sum_{n=1}^{\infty} \left( a_n \cos\left( \frac{n\pi x}{L} \right) + b_n \sin\left( \frac{n\pi x}{L} \right) \right) \right] dx
$$

La integral es lineal, por lo que se distribuye sobre la suma:

$$
= \frac{1}{2} a_0 \int_{-L}^{L} dx
+ \sum_{n=1}^{\infty} a_n \int_{-L}^{L} \cos\left( \frac{n\pi x}{L} \right) dx
+ \sum_{n=1}^{\infty} b_n \int_{-L}^{L} \sin\left( \frac{n\pi x}{L} \right) dx
$$


Las integrales de seno y coseno sobre el intervalo $ [-L, L] $ se anulan para  $ n \geq 1 $:

$$
\int_{-L}^{L} \cos\left( \frac{n\pi x}{L} \right) dx = 0, \quad \text{para } n \geq 1
$$

$$
\int_{-L}^{L} \sin\left( \frac{n\pi x}{L} \right) dx = 0, \quad \text{para todo } n
$$

Esto ocurre porque:

- $ \sin\left( \frac{n\pi x}{L} \right) $ es una función **impar**, su integral sobre $ [-L, L] $ se cancela por simetría.
- $ \cos\left( \frac{n\pi x}{L} \right) $ es **par**, pero su área positiva y negativa también se cancelan (salvo para $ n = 0 $).

Entonces, se tiene:

$$
\int_{-L}^{L} f(x) \, dx = \frac{1}{2} a_0 \cdot \int_{-L}^{L} dx = \frac{1}{2} a_0 \cdot (2L) = a_0 L
$$

Entonces:

$$
a_0 = \frac{1}{L} \int_{-L}^{L} f(x) \, dx
$$

Este es el coeficiente $ a_0 $ de la serie de Fourier.


Cálculo del coeficiente $ a_n $ de la serie de Fourier:

La fórmula base de la serie de Fourier es:

$$
f(x) = \frac{1}{2} a_0 + \sum_{m=1}^{\infty} \left( a_m \cos\left(\frac{m \pi x}{L}\right) + b_m \sin\left(\frac{m \pi x}{L}\right) \right)
$$

Multiplicamos ambos lados por $ \cos\left(\frac{n \pi x}{L}\right) $ integrando $ [-L, L] $:

$$
\int_{-L}^L f(x) \cos\left(\frac{n \pi x}{L}\right) dx = \int_{-L}^L \left[ \frac{1}{2} a_0 + \sum_{m=1}^\infty \left( a_m \cos\left(\frac{m \pi x}{L}\right) + b_m \sin\left(\frac{m \pi x}{L}\right) \right) \right] \cos\left(\frac{n \pi x}{L}\right) dx
$$

Por linealidad, separamos:

$$
= \frac{1}{2} a_0 \int_{-L}^L \cos\left(\frac{n \pi x}{L}\right) dx + \sum_{m=1}^\infty a_m \int_{-L}^L \cos\left(\frac{m \pi x}{L}\right) \cos\left(\frac{n \pi x}{L}\right) dx + \sum_{m=1}^\infty b_m \int_{-L}^L \sin\left(\frac{m \pi x}{L}\right) \cos\left(\frac{n \pi x}{L}\right) dx
$$



1. **Primer término:**

$$
\int_{-L}^L \cos\left(\frac{n \pi x}{L}\right) dx = 0 \quad \text{para todo } n \geq 1
$$

Se anula porque el coseno completo sobre un número entero de semiperíodos oscila simétricamente.

2. **Tercer término:**

$$
\int_{-L}^L \sin\left(\frac{m \pi x}{L}\right) \cos\left(\frac{n \pi x}{L}\right) dx = 0
$$

Se anula por ortogonalidad entre seno (función impar) y coseno (función par) en intervalo simétrico.

3. **Segundo término:**

$$
\sum_{m=1}^\infty a_m \int_{-L}^L \cos\left(\frac{m \pi x}{L}\right) \cos\left(\frac{n \pi x}{L}\right) dx
$$

Todos los terminos se eliminan, cuando $ m = n $ gracias a la ortogonalidad:

$$
\int_{-L}^L \cos\left(\frac{m \pi x}{L}\right) \cos\left(\frac{n \pi x}{L}\right) dx =
\begin{cases}
0, & m \neq n \\
L, & m = n
\end{cases}
$$


Entonces:

$$
\int_{-L}^L f(x) \cos\left(\frac{n \pi x}{L}\right) dx = a_n \cdot L \quad \Rightarrow \quad a_n = \frac{1}{L} \int_{-L}^L f(x) \cos\left(\frac{n \pi x}{L}\right) dx
$$


Cálculo de $ b_n $:

Multiplicamos por $ \sin\left( \frac{n\pi x}{L} \right) $ e integramos:

$$
\int_{-L}^{L} f(x) \sin\left( \frac{n\pi x}{L} \right) dx = \int_{-L}^{L} \left[ \frac{1}{2} a_0 + \sum_{m=1}^{\infty} \left( a_m \cos\left( \frac{m\pi x}{L} \right) + b_m \sin\left( \frac{m\pi x}{L} \right) \right) \right] \sin\left( \frac{n\pi x}{L} \right) dx
$$

- $ \int_{-L}^{L} a_0 \sin(...) = 0 $
- $ \int_{-L}^{L} a_m \cos(...) \sin(...) = 0 $
- Solo sobrevive el término $ b_n $:

$$
b_n \int_{-L}^{L} \sin^2\left( \frac{n\pi x}{L} \right) dx = b_n L
\Rightarrow
b_n = \frac{1}{L} \int_{-L}^{L} f(x) \sin\left( \frac{n\pi x}{L} \right) dx
$$
