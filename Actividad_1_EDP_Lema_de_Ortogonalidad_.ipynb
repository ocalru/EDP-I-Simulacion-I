Se demostrara si la funcion

$$\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right)\,dx $$

es igual a 0.

La función $ \cos\left(\frac{n\pi x}{L} \right) $ es par:

$$
\cos(-x) = \cos(x)
$$

Por lo tanto:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right)\,dx = 2 \int_0^L \cos\left(\frac{n\pi x}{L}\right)\,dx
$$


Sea:

$$
u = \frac{n\pi x}{L} \Rightarrow du = \frac{n\pi}{L} dx \Rightarrow dx = \frac{L}{n\pi} du
$$

Cuando $ x = 0 \Rightarrow u = 0 $

Cuando $ x = L \Rightarrow u = n\pi $

Entonces:

$$
\int_0^L \cos\left(\frac{n\pi x}{L}\right)\,dx = \int_0^{n\pi} \cos(u) \cdot \frac{L}{n\pi}\,du
= \frac{L}{n\pi} \int_0^{n\pi} \cos(u)\,du
$$



$$
\int_0^{n\pi} \cos(u)\,du = \sin(u)\big|_0^{n\pi} = \sin(n\pi) - \sin(0) = 0 - 0 = 0
$$

Por lo tanto:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right)\,dx = 2 \cdot \frac{L}{n\pi} \cdot 0 = 0
$$



$$
\int_{-L}^{L} \sin\left(\frac{n\pi x}{L}\right)\,dx = 0
$$

Se demostrara si la integral:



$$
\int_{-L}^{L} \sin\left(\frac{n\pi x}{L}\right)\,dx
$$

Es igual a $ 0 $


La función:

$$
f(x) = \sin\left(\frac{n\pi x}{L}\right)
$$

es una función **impar**, porque:

$$
f(-x) = \sin\left(-\frac{n\pi x}{L}\right) = -\sin\left(\frac{n\pi x}{L}\right) = -f(x)
$$

Para funciones impares $ f(x) $ integradas en intervalos simétricos $ [-a, a] $:

$$
\int_{-a}^{a} f(x)\,dx = 0
$$

Entonces:

$$
\int_{-L}^{L} \sin\left(\frac{n\pi x}{L}\right)\,dx = 0 \quad \text{(por simetría)}
$$



Se quiere demostrar que:

$$
\int_{-L}^{L} \cos\left(\frac{n \pi x}{L}\right) \cos\left(\frac{m \pi x}{L}\right) \, dx =
\begin{cases}
0, & \text{si } n \neq m, \\
L, & \text{si } n = m \neq 0,
\end{cases}
$$

Donde $ n, m \in \mathbb{Z}^+ $ (enteros positivos).


Se tiene que la integral:
$$
\int_{-L}^L \cos\left(\frac{n \pi x}{L}\right) \cos\left(\frac{m \pi x}{L}\right) dx
$$

Usando la **identidad trigonométrica**:

$$
\cos A \cos B = \frac{1}{2} \left[ \cos(A - B) + \cos(A + B) \right].
$$

Aplicamos esto con:

$$
A = \frac{n \pi x}{L}, \quad B = \frac{m \pi x}{L}.
$$

Entonces:

$$
\cos\left(\frac{n \pi x}{L}\right) \cos\left(\frac{m \pi x}{L}\right) = \frac{1}{2} \left[ \cos\left(\frac{(n - m)\pi x}{L}\right) + \cos\left(\frac{(n + m)\pi x}{L}\right) \right].
$$

La integral queda:

$$
\int_{-L}^L \cos\left(\frac{n \pi x}{L}\right) \cos\left(\frac{m \pi x}{L}\right) dx =
\frac{1}{2} \int_{-L}^L \cos\left(\frac{(n - m)\pi x}{L}\right) dx +
\frac{1}{2} \int_{-L}^L \cos\left(\frac{(n + m)\pi x}{L}\right) dx.
$$



Resolviendo la primera integral:

$$
I_1 = \frac{1}{2} \int_{-L}^{L} \cos\left(\frac{(n - m)\pi x}{L}\right) dx
$$

Encontrando la primitiva:

$$
\int \cos(ax)\, dx = \frac{1}{a} \sin(ax) + C
$$

Donde $ a = \frac{(n - m)\pi}{L} $

$$
\int \cos\left(\frac{(n - m)\pi x}{L}\right) dx = \frac{L}{(n - m)\pi} \sin\left(\frac{(n - m)\pi x}{L}\right)
$$

Evaluando entre $-L$ y $L$:**

$$
I_1 = \frac{1}{2} \cdot \left[ \frac{L}{(n - m)\pi} \sin\left(\frac{(n - m)\pi x}{L} \right) \right]_{-L}^{L}
$$

$$
= \frac{1}{2} \cdot \frac{L}{(n - m)\pi} \left[ \sin((n - m)\pi) - \sin(-(n - m)\pi) \right]
$$

$$
= \frac{1}{2} \cdot \frac{L}{(n - m)\pi} \cdot \left[ \sin((n - m)\pi) + \sin((n - m)\pi) \right] = \frac{L}{(n - m)\pi} \cdot \sin((n - m)\pi)
$$

$$
= \frac{L}{(n - m)\pi} \cdot 0 = 0 \quad \text{(porque } \sin(k\pi) = 0 \text{ para todo } k \in \mathbb{Z})
$$

Evaluando la segunda integral:

$$
I_2 = \frac{1}{2} \int_{-L}^{L} \cos\left(\frac{(n + m)\pi x}{L}\right) dx
$$

Calculando la primitiva:

$$
\int \cos\left(\frac{(n + m)\pi x}{L}\right) dx = \frac{L}{(n + m)\pi} \sin\left(\frac{(n + m)\pi x}{L} \right)
$$

Evaluando entre $-L$ y $ L $:**

$$
I_2 = \frac{1}{2} \cdot \left[ \frac{L}{(n + m)\pi} \sin\left(\frac{(n + m)\pi x}{L} \right) \right]_{-L}^{L}
$$

$$
= \frac{1}{2} \cdot \frac{L}{(n + m)\pi} \left[ \sin((n + m)\pi) - \sin(-(n + m)\pi) \right]
$$

$$
= \frac{1}{2} \cdot \frac{L}{(n + m)\pi} \left[ \sin((n + m)\pi) + \sin((n + m)\pi) \right]
= \frac{L}{(n + m)\pi} \cdot \sin((n + m)\pi)
$$

$$
= \frac{L}{(n + m)\pi} \cdot 0 = 0 \quad \text{(porque } \sin(k\pi) = 0 \text{ para todo } k \in \mathbb{Z})
$$
Se tiene que:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right) \cos\left(\frac{m\pi x}{L}\right) dx = 0 \quad \text{para } n \ne m
$$


Caso especial: $ n = m $


Partimos de la integral original:

$$
\int_{-L}^{L} \cos\left(\frac{n\pi x}{L}\right) \cos\left(\frac{m\pi x}{L}\right) dx
$$

En el caso especial en que $ n = m $, esta expresión se convierte en:

$$
\int_{-L}^{L} \cos^2\left(\frac{n\pi x}{L}\right) dx
$$

Esto ocurre porque estamos multiplicando una función consigo misma. Es decir, al ser $ n = m $, ambas funciones son idénticas, y la integral se convierte en la del cuadrado de una sola función:

$$
\cos\left(\frac{n\pi x}{L}\right) \cdot \cos\left(\frac{n\pi x}{L}\right) = \cos^2\left(\frac{n\pi x}{L}\right)
$$

Por lo tanto, la integral queda:

$$
\int_{-L}^{L} \cos^2\left(\frac{n\pi x}{L}\right) dx
$$




Usamos la identidad trigonométrica:

$$
\cos^2(\theta) = \frac{1 + \cos(2\theta)}{2}
$$

Entonces:

$$
\int_{-L}^{L} \cos^2\left(\frac{n\pi x}{L}\right) dx = \int_{-L}^{L} \frac{1 + \cos\left(\frac{2n\pi x}{L}\right)}{2} dx
$$

$$
= \frac{1}{2} \int_{-L}^{L} 1 \, dx + \frac{1}{2} \int_{-L}^{L} \cos\left(\frac{2n\pi x}{L}\right) dx
$$



**Primera parte:**

$$
\frac{1}{2} \int_{-L}^{L} 1 \, dx = \frac{1}{2} \cdot [x]_{-L}^{L} = \frac{1}{2} (L - (-L)) = \frac{1}{2} \cdot 2L = L
$$

**Segunda parte**

$$
\frac{1}{2} \int_{-L}^{L} \cos\left(\frac{2n\pi x}{L}\right) dx
$$

La primitiva es:

$$
\int \cos(ax) \, dx = \frac{1}{a} \sin(ax)
\quad \text{con } a = \frac{2n\pi}{L}
$$

Entonces:

$$
\int \cos\left(\frac{2n\pi x}{L}\right) dx = \frac{L}{2n\pi} \sin\left(\frac{2n\pi x}{L}\right)
$$

Evaluando:

$$
\left[ \frac{L}{2n\pi} \sin\left(\frac{2n\pi x}{L}\right) \right]_{-L}^{L}
= \frac{L}{2n\pi} \left[ \sin(2n\pi) - \sin(-2n\pi) \right] = \frac{L}{2n\pi} (0 + 0) = 0
$$

Entonces se tiene que:

$$
\int_{-L}^{L} \cos^2\left(\frac{n\pi x}{L}\right) dx = L
$$
