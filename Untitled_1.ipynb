# Método de Aceptación y Rechazo para una Distribución Triangular

## Distribución Triangular
La distribución triangular es una distribución continua definida por tres parámetros:
- $a$: límite inferior,
- $b$: límite superior,
- $c$: moda, tal que $a < c < b$.

Su función de densidad de probabilidad $f(x)$ está dada por:

$$
f(x) =
\begin{cases}
0, & x < a \\
\frac{2(x - a)}{(b - a)(c - a)}, & a \le x < c \\
\frac{2}{b - a}, & x = c \\
\frac{2(b - x)}{(b - a)(b - c)}, & c < x \le b \\
0, & x > b
\end{cases}
$$

En nuestro caso:  
- $a = 1$  
- $c = 3$  
- $b = 5$


```python
# Simulación por el método de Aceptación y Rechazo para una distribución Triangular
# Parámetros: a = 1, b = 3, c = 5

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import triang

# Definimos los parámetros de la distribución triangular
A = 1      # límite inferior
B = 5      # límite superior
C = 3      # moda (el valor más probable)

# Altura máxima de la función de densidad (valor máximo de f(x))
M = 2 / (B - A)  # Este valor se usa como cota superior en la aceptación/rechazo

# Cantidad de muestras que deseamos generar
n_samples = 10000

# Lista vacía donde se almacenarán los valores aceptados
samples = []

# Definimos la función de densidad de la distribución triangular
# Esta función evalúa f(x) para un valor dado de x

def triangular_pdf(x, a, c, b):
    if x < a or x > b:
        return 0  # f(x) es 0 fuera del intervalo [a, b]
    elif x < c:
        # Lado izquierdo de la distribución triangular
        return 2 * (x - a) / ((b - a) * (c - a))
    elif x == c:
        # Altura máxima en la moda
        return 2 / (b - a)
    else:
        # Lado derecho de la distribución triangular
        return 2 * (b - x) / ((b - a) * (b - c))
```

## Método de Aceptación y Rechazo
Este método genera variables aleatorias de una distribución arbitraria $f(x)$, utilizando una distribución uniforme como propuesta.

**Pasos**:
1. Generar $x \sim U(a, b)$.
2. Generar $y \sim U(0, M)$, donde $M$ es el valor máximo de $f(x)$, que en la triangular es $\frac{2}{b - a}$.
3. Si $y \le f(x)$, se acepta el valor $x$; de lo contrario, se rechaza.

Para la distribución triangular con $a = 1$, $c = 3$, $b = 5$, el valor máximo de la densidad es:

$$
M = \frac{2}{b - a} = \frac{2}{4} = 0.5
$$

```python

# Generamos las muestras utilizando el método de Aceptación y Rechazo
# Repetimos el proceso hasta recolectar el número deseado de muestras
np.random.seed(0)  # Fijamos semilla para reproducibilidad

# Bucle principal de generación
# Mientras el número de muestras aceptadas sea menor que el deseado:
while len(samples) < n_samples:
    # Paso 1: generar una muestra candidata x uniformemente entre a y b
    x = np.random.uniform(A, B)

    # Paso 2: generar un valor y uniformemente entre 0 y M
    # M es la cota superior de la función de densidad f(x)
    y = np.random.uniform(0, M)

    # Paso 3: aceptar x si y está por debajo de la curva de f(x)
    if y <= triangular_pdf(x, A, C, B):
        samples.append(x)  # x es aceptado y se guarda como muestra válida

# Convertimos la lista de muestras a un arreglo de NumPy para cálculos estadísticos
samples = np.array(samples)
```




## Estadísticas Teóricas
- **Media**:
$$
\mu = \frac{a + b + c}{3} = \frac{1 + 3 + 5}{3} = 3.0
$$

- **Varianza**:
$$
\sigma^2 = \frac{a^2 + b^2 + c^2 - ab - ac - bc}{18} = \frac{1 + 25 + 9 - 5 - 3 - 15}{18} = \frac{12}{18} = 0.6667
$$

- **Desviación estándar**:
$$
\sigma = \sqrt{\sigma^2} = \sqrt{0.6667} \approx 0.8165
$$

Estas estadísticas teóricas sirven como referencia para evaluar la calidad de la simulación.

```python
# Cálculos estadísticos empíricos a partir de las muestras simuladas
empirical_mean = np.mean(samples)  # Media observada
empirical_std = np.std(samples)    # Desviación estándar observada

# Cálculos teóricos de la media y desviación estándar para distribución triangular
mean_theoretical = (A + B + C) / 3
std_theoretical = np.sqrt((A**2 + B**2 + C**2 - A*B - A*C - B*C) / 18)

# Imprimimos los resultados estadísticos
print("Resultados estadísticos:")
print(f"Media empírica: {empirical_mean:.4f}")
print(f"Desviación estándar empírica: {empirical_std:.4f}")
print(f"Media teórica: {mean_theoretical:.4f}")
print(f"Desviación estándar teórica: {std_theoretical:.4f}")

# Graficamos histograma de las muestras con la función de densidad teórica superpuesta
x_vals = np.linspace(A, B, 1000)

# Evaluamos la función de densidad teórica en los puntos x_vals
y_vals = [triangular_pdf(x, A, C, B) for x in x_vals]

plt.figure(figsize=(10, 6))
plt.hist(samples, bins=50, density=True, alpha=0.6, color='skyblue', label='Histograma de muestras')
plt.plot(x_vals, y_vals, 'r-', lw=2, label='PDF Teórica')
plt.title('Distribución Triangular por Aceptación y Rechazo')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.legend()
plt.grid(True)
plt.show()
```
