```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Datos
X = np.array([6.8, 3.3, 4.1, 4.2, 4.8, 3.9, 4.9, 7.7, 3.1, 5.9, 5]).reshape(-1, 1)
Y = np.array([275, 142, 168, 197, 215, 188, 241, 295, 125, 266, 200])

# Crear y ajustar el modelo
modelo = LinearRegression()
modelo.fit(X, Y)

# Coeficientes
pendiente = modelo.coef_[0]
interseccion = modelo.intercept_

print(f"Modelo: Y = {pendiente:.2f}X + {interseccion:.2f}")

# Predicciones
Y_pred = modelo.predict(X)

# Graficar
plt.scatter(X, Y, color='blue', label='Datos reales')
plt.plot(X, Y_pred, color='red', label='Recta de regresión')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión Lineal')
plt.legend()
plt.grid(True)
plt.show()
```
