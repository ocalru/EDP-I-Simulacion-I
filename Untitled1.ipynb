```python
# Movim

import numpy as np
import matplotlib.pyplot as plt

# =========================
# Parmetros
# =========================
T = 1.0
N_coarse = 64        # malla gruesa: N puntos (incrementos)
refine_factor = 8    # refinamiento: multiplica N por este factor
seed = 42

rng = np.random.default_rng(seed)

# =========================
# Funcin: trayectoria de BM en malla uniforme
# =========================
def brownian_path(T, N, rng):
    dt = T / N
    dW = np.sqrt(dt) * rng.standard_normal(N)
    W = np.concatenate([[0.0], np.cumsum(dW)])  # W_0=0
    t = np.linspace(0, T, N + 1)
    return t, W

# =========================
# Simulacin: malla gruesa
# =========================
t_c, W_c = brownian_path(T, N_coarse, rng)

# =========================
# Refinamiento: genera una trayectoria fina INDEPENDIENTE
# (y la superpone para mostrar el refinamiento)
# =========================
N_fine = N_coarse * refine_factor
t_f, W_f = brownian_path(T, N_fine, rng)

# =========================
# Grfica: comparacin
# =========================
plt.figure(figsize=(10,5))
plt.plot(t_f, W_f, linewidth=1.2, label=f"Refinada (N={N_fine})")
plt.plot(t_c, W_c, marker="o", markersize=3, linewidth=1.0, label=f"Gruesa (N={N_coarse})")
plt.title("Movimiento Browniano en [0,1] con refinamiento de malla")
plt.xlabel("t")
plt.ylabel("W(t)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

# =========================
# (Opcional) Refinamiento CONSISTENTE: misma trayectoria, ms puntos
# Construye una BM fina y toma submuestra para la gruesa
# =========================
rng2 = np.random.default_rng(seed)  # misma semilla para reproducibilidad
t_f2, W_f2 = brownian_path(T, N_fine, rng2)
W_c_from_fine = W_f2[::refine_factor]
t_c_from_fine = t_f2[::refine_factor]

plt.figure(figsize=(10,5))
plt.plot(t_f2, W_f2, linewidth=1.2, label=f"Refinada consistente (N={N_fine})")
plt.plot(t_c_from_fine, W_c_from_fine, marker="o", markersize=3, linewidth=1.0,
         label=f"Submuestra (equiv. N={N_coarse})")
plt.title("BM consistente: refinada y su submuestra en la malla gruesa")
plt.xlabel("t")
plt.ylabel("W(t)")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()
```
