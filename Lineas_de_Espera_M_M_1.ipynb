```python
from random import expovariate
from statistics import mean
from math import inf

# Parámetros
lamda = 4
mu    = 6
Num_Pkts = 100

count = 0          # paquetes que han salido
t = 0.0
N = 0              # número de clientes en el sistema

tll = expovariate(lamda)  # tiempo de la siguiente llegada
ts  = inf                 # tiempo de la siguiente salida (no hay al inicio)

t_evento_anterior = 0.0

tll_Data = []   # tiempos de llegada
ts_Data  = []   # tiempos de salida
r_Data   = []   # retrasos de cada cliente
Area     = []   # para L promedio

# PARA PROBABILIDADES P{N = n}
tiempo_en_estado = {}   # diccionario: estado n -> tiempo total en ese estado

# ---------------- Simulación ----------------
while count < Num_Pkts:
    if tll < ts:   # siguiente evento: llegada
        t = tll
        tll_Data.append(t)

        # tiempo transcurrido en el estado N actual
        dt = t - t_evento_anterior
        Area.append(dt * N)
        tiempo_en_estado[N] = tiempo_en_estado.get(N, 0.0) + dt
        t_evento_anterior = t

        # se procesa la llegada
        N += 1
        tll = t + expovariate(lamda)
        if N == 1:
            ts = t + expovariate(mu)

    else:          # siguiente evento: salida
        t = ts
        ts_Data.append(t)

        # tiempo transcurrido en el estado N actual
        dt = t - t_evento_anterior
        Area.append(dt * N)
        tiempo_en_estado[N] = tiempo_en_estado.get(N, 0.0) + dt
        t_evento_anterior = t

        # se procesa la salida
        N -= 1
        count += 1
        if N > 0:
            ts = t + expovariate(mu)
        else:
            ts = inf

# ---------------- Métricas L y W ----------------
for i in range(Num_Pkts):
    d = ts_Data[i] - tll_Data[i]
    r_Data.append(d)

print("El retraso promedio W es:", round(mean(r_Data), 4))
print("El número promedio de clientes L es:", round(sum(Area) / t, 4))

# ---------------- Probabilidades P{N = n} ----------------
T_total = sum(tiempo_en_estado.values())  # tiempo total simulado (≈ t)

print("\nProbabilidades estimadas de que haya n clientes en el sistema:")
for estado in sorted(tiempo_en_estado.keys()):
    p_n = tiempo_en_estado[estado] / T_total
    print(f"P(N = {estado}) ≈ {p_n:.4f}")

```
