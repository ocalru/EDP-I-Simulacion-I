Método de las 12 Uniformes para Generar una Distribución Normal Estándar


Una variable aleatoria $ U $ se dice que tiene una **distribución uniforme continua en el intervalo** $ [0,1] $ —denotado $ U \sim \text{Unif}(0,1) $— si **todos los valores del intervalo tienen la misma probabilidad "relativa"** de ser tomados. Matemáticamente, esto se expresa mediante su **función de densidad de probabilidad** (PDF).




La función de distribución acumulada (FDA) de $ U $ se define como:

$$
F_U(u) = \mathbb{P}(U \leq u)
$$

En el caso de la distribución uniforme continua en el intervalo $ [0,1] $, esta función es:

$$
F_U(u) =
\begin{cases}
0, & \text{si } u < 0 \\
u, & \text{si } 0 \leq u \leq 1 \\
1, & \text{si } u > 1
\end{cases}
$$

Este resultado se justifica porque, en una distribución uniforme continua en $ [0,1] $, la probabilidad de caer en un subintervalo $ [0,u] $ para $ u \in [0,1] $ es simplemente el **largo del subintervalo**, es decir, $ u $.


La **función de densidad de probabilidad (PDF)** de una variable aleatoria continua se define como:

$$
f_U(u) = \frac{d}{du} F_U(u)
$$

Entonces se deriva cada tramo de la funcion acumulada:

- Si $ u < 0 $: $ F_U(u) = 0 \Rightarrow f_U(u) = 0 $
- Si $ 0 < u < 1 $: $ F_U(u) = u \Rightarrow f_U(u) = 1 $
- Si $ u > 1 $: $ F_U(u) = 1 \Rightarrow f_U(u) = 0 $

Por lo tanto, la función de densidad de $ U $ es:

$$
f_U(u) =
\begin{cases}
1, & \text{si } 0 \leq u \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$

La función $ f_U(u) $ nos dice la **densidad de probabilidad** en cada punto. En este caso:

- Todos los valores del intervalo $[0,1]$ son **igual de probables** (la densidad es constante).
- Fuera del intervalo, la densidad es cero: la variable nunca toma esos valores.


Por definición de función de densidad:

$$
f_U(u) \geq 0 \quad \text{para todo} u \in \mathbb{R}
$$

En este caso:

$$
f_U(u) =
\begin{cases}
1, & \text{si } 0 \leq u \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$

Luego, claramente $ f_U(u) \geq 0 \quad \forall\, u \in \mathbb{R} $.


La segunda propiedad que debe cumplir una función de densidad es que:

$$
\int_{-\infty}^{\infty} f_U(u)\,du = 1
$$

Desarrollamos esta integral usando la definición por tramos:

$$
\int_{-\infty}^{\infty} f_U(u)\,du =
\int_{-\infty}^0 f_U(u)\,du +
\int_0^1 f_U(u)\,du +
\int_1^{\infty} f_U(u)\,du
$$

Sustituimos valores:

$$
= \int_{-\infty}^0 0 \,du + \int_0^1 1 \,du + \int_1^{\infty} 0 \,du
= 0 + \left[u\right]_0^1 + 0 = 1 - 0 = 1
$$

Se cumple que la integral total de la densidad es igual a 1, como exige la teoría de probabilidad.

    <font size="10" color="lightblue"> Método de las 12 Uniformes para Generar una Distribución Normal Estándar. </font>

El Método de las 12 Uniformes es una técnica clásica de generación de variables aleatorias normales estándar $ Z \sim \mathcal{N}(0,1) $ a partir de 12 variables aleatorias independientes con distribución uniforme $ U_i \sim \text{Unif}(0,1) $. A continuación, se desarrolla formalmente el método.


Sea $ U \sim \text{Unif}(0,1) $, es decir, una variable aleatoria continua con función de distribución acumulada $ F_U(u) $ definida por:

$$
F_U(u) =
\begin{cases}
0, & \text{si } u < 0 \\
u, & \text{si } 0 \leq u \leq 1 \\
1, & \text{si } u > 1
\end{cases}
$$

La función de densidad $ f_U(u) $ se obtiene como la derivada de $ F_U(u) $:

$$
f_U(u) = \frac{d}{du} F_U(u) =
\begin{cases}
1, & \text{si } 0 \leq u \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$



Sea $ U_1, U_2, \dots, U_{12} \sim \text{i.i.d. Unif}(0,1) $, es decir, son **variables aleatorias independientes e idénticamente distribuidas** con distribución uniforme continua en el intervalo $ [0,1] $.

Esto implica que:

- Cada $ U_i $ tiene la **misma distribución**, es decir, la misma función de densidad:
  $$
  f_{U_i}(u) =
  \begin{cases}
  1, & \text{si } 0 \leq u \leq 1 \\
  0, & \text{en otro caso}
  \end{cases}
  $$

- Las variables son **mutuamente independientes**, es decir, conocer el valor de una no da información sobre las demás.



Se define la variable aleatoria $ S_{12} $ como la **suma de 12 variables aleatorias continuas independientes e idénticamente distribuidas** con distribución uniforme en el intervalo \([0,1]\).

$$
S_{12} := \sum_{i=1}^{12} U_i
$$



- Esperanza de $ S_{12} $


Sabemos que para una variable aleatoria continua $ U_i \sim \text{Unif}(0,1) $, la esperanza se define como:

$$
\mathbb{E}[U_i] = \int_{-\infty}^{\infty} u \cdot f_{U_i}(u) \, du
$$

Donde:

- $ f_{U_i}(u) $ es la función de densidad de probabilidad de $ U_i $. Como $ U_i \sim \text{Unif}(0,1) $, se tiene:

$$
f_{U_i}(u) =
\begin{cases}
1, & \text{si } 0 \leq u \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$

Sustituyendo en la definición de esperanza:

$$
\mathbb{E}[U_i] = \int_0^1 u \cdot 1 \, du = \int_0^1 u \, du
$$

Resolviendo la integral:

$$
\int_0^1 u \, du = \left[ \frac{u^2}{2} \right]_0^1 = \frac{1^2}{2} - \frac{0^2}{2} = \frac{1}{2}
$$

Por lo tanto:

$$
\mathbb{E}[U_i] = \frac{1}{2}
$$



Sabemos que cada $ U_i \sim \text{Unif}(0,1) $ tiene esperanza:

$$
\mathbb{E}[U_i] = \frac{1}{2}
$$

Entonces, usando la propiedad de linealidad de la esperanza:

$$
\mathbb{E}[S_{12}] = \mathbb{E}\left[ \sum_{i=1}^{12} U_i \right] = \sum_{i=1}^{12} \mathbb{E}[U_i]
$$

Sustituimos el valor de $ \mathbb{E}[U_i] $:

$$
\mathbb{E}[S_{12}] = \sum_{i=1}^{12} \frac{1}{2} = 12 \cdot \frac{1}{2} = \frac{12}{2} = 6
$$



- Varianza de $ S_{12} $

 Las variables son independientes, por lo que la varianza de la suma es la suma de las varianzas:

$$
\text{Var}(S_{12}) = \sum_{i=1}^{12} \text{Var}(U_i)
$$

La varianza de una variable uniforme $ U_i \sim \text{Unif}(0,1) $ es:

$$
\text{Var}(U_i) = \mathbb{E}[U_i^2] - (\mathbb{E}[U_i])^2
$$

Calculamos $ \mathbb{E}[U_i^2] $:

Por definición, la esperanza de $ U_i^2 $ es:

$$
\mathbb{E}[U_i^2] = \int_{-\infty}^{\infty} u^2 \cdot f_{U_i}(u) \, du
$$

Dado que $ U_i \sim \text{Unif}(0,1) $, su densidad es:

$$
f_{U_i}(u) =
\begin{cases}
1, & \text{si } 0 \leq u \leq 1 \\
0, & \text{en otro caso}
\end{cases}
$$

Entonces la integral se reduce a:

$$
\mathbb{E}[U_i^2] = \int_0^1 u^2 \cdot 1 \, du = \int_0^1 u^2 \, du
$$

Calculamos la integral:

$$
\int_0^1 u^2 \, du = \left[ \frac{u^3}{3} \right]_0^1 = \frac{1^3}{3} - \frac{0^3}{3} = \frac{1}{3}
$$

Por lo tanto:

$$
\boxed{ \mathbb{E}[U_i^2] = \frac{1}{3} }
$$

Entonces:

$$
\text{Var}(U_i) = \frac{1}{3} - \left(\frac{1}{2}\right)^2 = \frac{1}{3} - \frac{1}{4} = \frac{4 - 3}{12} = \frac{1}{12}
$$

Sabemos que:

$$
\text{Var}(S_{12}) = \sum_{i=1}^{12} \text{Var}(U_i)
$$

Sustituyendo $ \text{Var}(U_i) = \frac{1}{12} $ en $ \text{Var}(S_{12}) $:

$$
\text{Var}(S_{12}) = \sum_{i=1}^{12} \frac{1}{12} = 12 \cdot \frac{1}{12} = 1
$$

$$
\boxed{ \text{Var}(S_{12}) = 1 }
$$

Se define la variable:

$$
Z := S_{12} - 6
$$


Sabemos que la esperanza de una variable aleatoria $ X $ desplazada por una constante $ c \in \mathbb{R} $ es:

$$
\mathbb{E}[X - c] = \mathbb{E}[X] - c
$$

Aplicando esta propiedad a $ Z = S_{12} - 6 $:

$$
\mathbb{E}[Z] = \mathbb{E}[S_{12} - 6] = \mathbb{E}[S_{12}] - 6
$$

Ya habíamos calculado que:

$$
\mathbb{E}[S_{12}] = 6
$$

Entonces:

$$
\mathbb{E}[Z] = 6 - 6 = 0
$$

$$
\boxed{ \mathbb{E}[Z] = 0 }
$$

Sabemos que si $ X $ es una variable aleatoria y $ c \in \mathbb{R} $, entonces:

$$
\text{Var}(X - c) = \text{Var}(X)
$$

Esto se debe a que la varianza es invariante ante desplazamientos constantes.

Entonces:

$$
\text{Var}(Z) = \text{Var}(S_{12} - 6) = \text{Var}(S_{12})
$$

Y como ya habíamos obtenido:

$$
\text{Var}(S_{12}) = 1
$$

Entonces:

$$
\boxed{ \text{Var}(Z) = 1 }
$$




```python
import numpy as np # Se importa la libreria numpy como np
import matplotlib.pyplot as plt # Se importa la libreria matplotlib.pyplot como plt

# Parámetros de la normal deseada
mu = 10         # Media deseada
sigma = 2       # Desviación estándar deseada (√varianza)

# Número de muestras
n_samples = 10000

# Método de las 12 uniformes
def generar_normal(mu, sigma, n):

    # Paso 1: generar matriz de tamaño (n, 12) con variables uniformes en (0, 1)
    U = np.random.uniform(0, 1, size=(n, 12))

    # Paso 2: sumar las 12 variables por fila
    S_12 = np.sum(U, axis=1)

    # Paso 3: restar 6 para obtener normal estándar aproximada
    Z = S_12 - 6

    # Paso 4: transformar a normal con media mu y desviación estándar sigma
    X = mu + sigma * Z
    return X

```

```python
# Generar datos
datos = generar_normal(mu, sigma, n_samples)

# Graficar histograma
plt.figure(figsize=(8, 4))
plt.hist(datos, bins=50, density=True, alpha=0.6, color='skyblue', edgecolor='black')
plt.title(f'Distribución generada: Normal({mu}, {sigma**2})')
plt.xlabel('Valor')
plt.ylabel('Densidad')
plt.grid(True)
plt.show()
```
