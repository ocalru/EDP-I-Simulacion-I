<p align="center">
    <font size="10" color="lightblue" >Variables de Control</font>
</p>

Es una técnica que se usa en simulación y estimación estadística para reducir la varianza de un estimador, sin introducir sesgo.

Se define la variable $ Z = X + c(Y - μ_Y)$, esta variable se usa para estimar la un valor $I$ usando un estimador propuesto $Z$ que depende de las variables aleatorias $X,Y$

- Se busca que el estimador $Z$ tenga la misma esperenza que la variable aleatoria $X$:
$$\mathbb{E}(X) = \mathbb{E}(Z)$$
Esto se logra haciendo  $\; \mathbb{E}(Y - μ_Y) = 0 ⇒ \mathbb{E}(Y) = μ_Y$, entonces debemos de conocer la esperanza de la variable aleatoria $Y$.


- La correlacion entre $ Y, X $ debe ser diferente de $0\;$( $\mathbb{Cov}(Y,X) \neq 0 ⇒ (X, Y) \;$sean variables dependientes).
Si hay correlación, al ajustar $ X $ , podemos cancelar parte de la variabilidad de $ X $.
Esto hace que:

$$
\mathbb{Var}(Z) < \mathbb{Var}(X)
$$

- Si $ c = 0 \;$ obtenemos $ Z = X $

- Si $ c $ se elige **óptimamente** (en función de la covarianza entre $ X $ y $ Y $, se obtiene la **mínima varianza posible**.




<p align="center">
    <font size="5" color="lightblue" >Calculo de la esperanza de $Z$ </font>
</p>

Queremos verificar que:

$$
\mathbb{E}(Z) = \mathbb{E}(X)
$$

Partimos de:

$$
Z = X + c (Y - \mu_Y)
$$

Aplicamos la esperanza:

$$
\mathbb{E}(Z) = \mathbb{E}[X + c (Y - \mu_Y)]
$$

Usando la linealidad de la esperanza:

$$
\mathbb{E}(Z) = \mathbb{E}(X) + c \, (\mathbb{E}(Y) - \mu_Y)
$$

Como $\mathbb{E}(Y) = \mu_Y$, el término se anula:

$$
\mathbb{E}(Z) = \mathbb{E}(X)
$$

Por lo tanto, $ Z $ es un **estimador insesgado** de $ X $


<p align="center">
    <font size="5" color="lightblue" >Calculo de la varianza de $Z$ </font>
</p>

Sabemos que:

$$
\mathrm{Var}(Z) = \mathrm{Var}(X + c (Y - \mu_Y))
$$

Como $ \mu_Y $ es constante:

$$
\mathrm{Var}(Z) = \mathrm{Var}(X + cY - c\mu_Y) = \mathrm{Var}(X + cY)
$$

Usando propiedades de la varianza y omitiendo los calculos hechos se tiene:

$$
\mathrm{Var}(Z) = \mathrm{Var}(X) + c^2 \mathrm{Var}(Y) + 2c \, \mathrm{Cov}(X, Y)
$$

La varianza de $ Z $ depende de la varianza de $ X $, la de $ Y $, y su covarianza.


Si tomamos $ c = 0 $:

$$
\mathrm{Var}(Z) = \mathrm{Var}(X)
$$

No hay mejora.  
Pero si elegimos $ c $ óptimo, **podemos reducir la varianza**.

Derivamos respecto a $ c \;$ usando el criterio de la primer derivada para encontrar el optimo:

$$
\frac{d}{dc} \, \mathrm{Var}(Z) = 2c \, \mathrm{Var}(Y) + 2 \, \mathrm{Cov}(X, Y) = 0
$$

Omitiendo calculos se encuentra que:

$$
c^* = -\frac{\mathrm{Cov}(X, Y)}{\mathrm{Var}(Y)}
$$


Sustituimos $ c^* $ en la expresión de la varianza:

$$
\mathrm{Var}(Z) = \mathrm{Var}(X) + c^2 \mathrm{Var}(Y) + 2c \, \mathrm{Cov}(X, Y)
$$

Con $ c^* = -\dfrac{\mathrm{Cov}(X, Y)}{\mathrm{Var}(Y)} $:

$$
\mathrm{Var}(Z_{c^*}) = \mathrm{Var}(X) +
\left( \frac{\mathrm{Cov}(X, Y)^2}{\mathrm{Var}(Y)^2} \right) \mathrm{Var}(Y)
- 2 \left( \frac{\mathrm{Cov}(X, Y)}{\mathrm{Var}(Y)} \right) \mathrm{Cov}(X, Y)
$$

Simplificando:

$$
\mathrm{Var}(Z_{c^*}) = \mathrm{Var}(X)
- \frac{\mathrm{Cov}(X, Y)^2}{\mathrm{Var}(Y)}
$$

Por lo tanto, al usar $ c^* $, la varianza disminuye en el término
$$\dfrac{\mathrm{Cov}(X, Y)^2}{\mathrm{Var}(Y)}$$

- Para que la varianza de $Z$ disminuya, $\mathbb{Cov}(X,Y) → ∞$, no importa si las variables $X,Y$ tengan una correlacion negativa o positiva, lo importante es aumentar la magnitud de esa correlacion. ($ |\mathbb{Cov}(X,Y)| → ∞ $)


<p align="center">
    <font size="5" color="lightblue" > Ejemplo: </font>
</p>


Se tiene la integral definida $I$ en el intervalo $[a,b]$:

$$
I = \int_a^b g(x)\,dx
$$

Suponiendo que:

- $0 \leq g(x) \leq c$

- $a \leq x \leq b$

Como ejemplo, calculemos la integral:

$$
I = \int_0^1 \frac{1}{1 + x^2} \; dx
$$

Donde:

- $ g(x) = \frac{1}{1 + x^2} $
- $ a = 0 $
- $ b = 1 $

Tenemos que aproximamos el valor de la integral $I$:

$$
I = \int_a^b g(x)\, dx \;\;\approx\;\; (b-a)\cdot \frac{1}{n}\sum_{i=1}^n g(X_i).
$$
Tenemos el estimador:
$$
\boxed {
  \theta = (b-a)\frac{1}{n}\sum_{i=1}^n g(X_i)}
$$

Para estimar el valor de $I$ usando $\theta_2$:

$$
I = \frac{1}{n} \sum_{i=1}^{n} g(X_i)
= \frac{b-a}{n} \sum_{i=1}^{n} \frac{1}{1 + X_i^2}
$$

donde $ X_i \sim U(a,b) $ son variables **independientes** y **uniformemente distribuidas** en $[a,b]$.


Entonces el estimador insesgado para el valor de $I$ se define como:

$$ \theta= \frac{b-a}{n} \sum_{i=1}^{n} \frac{1}{1 + X_i^2}
$$






Queremos reducir la varianza del estimador $\theta$, usando variables de control, proponemos
$$
Y = \frac{X_1 + X_2 + \cdots + X_n}{n}
$$

Es decir, $ Y $ es el **promedio de los mismos** $ X_i $ que ya se usaron para calcular $ \theta $.

Elegimos $Y$ ya que cada $ X_i \sim U(a,b) $, por lo que:

$$
\mathbb{E}(X_i) = \frac{1}{b-a}
$$

Por linealidad de la esperanza:

$$
\mathbb{E}(Y) = \frac{1}{n} \sum_{i=1}^{n} \mathbb{E}(X_i)
= \frac{1}{n} \cdot n \cdot \frac{1}{b-a} = \frac{1}{b-a}
$$

Entonces conocemos su media ( $\mu_Y$ ) uno de los requisitos para aplicar al variable de control $Z$.


**¿Qué relación tiene con $ \theta $?**

Ambos dependen de los mismos $ X_i $.

- Si los $ X_i $ son grandes en promedio, $ Y $ también será grande.
$$\text {Si}\;  (X_1, X_2, \dots X_n) → ∞  ⇒ (Y → ∞) $$


Pero los términos $ \frac{1}{1 + X_i^2} $ tenderán a ser **más pequeños**, ya que el denominador aumenta.

$$\text {Si}\;  (\frac{1}{(1 + X_1^2)}, \frac{1}{(1 + X_2^2)}, \dots \frac{1}{(1 + X_n^2)}) → ∞  ⇒ (\theta → 0) $$

$$
\\
$$

$$
\Rightarrow \text{Existe una correlación negativa entre } \theta \text{ y } Y (\;\mathbb{Cov}(\theta, Y) < 0 )
$$

Esa correlación negativa es precisamente otra de las condiciones para poder aplicar la variable de control $Z$.

Definimos el nuevo estimador:

$$
Z = \theta + c \cdot(Y - \mu_Y)
$$


Calculando la esperenza, para ver si $Z$ es un estimador insesgado de $\theta$:

$$
\mathbb{E}(Z)  =\mathbb{E} [\theta + c \cdot(Y - \mu_Y) ]
$$

Aplicamos la **linealidad de la esperanza**:
$$
\mathbb{E}(Z) = \mathbb{E}[\theta] + \mathbb{E}[c \cdot (Y - \mu_Y)]
$$

Desarrollando:

$$
\mathbb{E}(Z) = \mathbb{E}[\theta] + c \, (\mathbb{E}[Y] - \mu_Y)
$$

$$
   \boxed{\mathbb{E}(Z) = \mathbb{E}[\theta]}
$$


Entonces:

$$
   \boxed{\mathbb{E}(Z) = \mathbb{E}[\theta] = I}
$$

Por lo tanto, $ Z $ es un **estimador insesgado** de $ \theta $. Y a su vez es un estimador insesgado de el valor de la integral $I$.

No se modifica el **valor esperado**, que sigue siendo $ \theta $.

Se tiene que la varianza de $Z$ es:

$$
\boxed{
\mathrm{Var}(Z) = \mathrm{Var}(\theta)
- \frac{\mathrm{Cov}(\theta , Y)^2}{\mathrm{Var}(Y)}
}
$$

Si $ | \mathrm{Cov}(\theta , Y)|  → ∞  $ entonces $\mathbb{Var}(Z) → 0 $ , la varianza de $Z$ baja y se cumple que:

$$
\boxed {\mathrm{Var}(Z) < \mathrm{Var}(\theta)}
$$

<p align="center">
    <font size="5" color="lightblue" > Aproximacion de los valores $\mathrm{Var}(\theta), \mathrm{Cov}(\theta , Y) ,\mathrm{Var}(Y) $ </font>
</p>



Se pueden calcular los valores teoricos de $ \mathrm{Var}(\theta) ,\mathrm{Var}(Y) , \mathrm{Cov}(X, Y)$ pero al ser demasiado complicados de calcular, usamos sus versiones muestrales:

$$
S_{\theta}^2 = \frac{1}{n - 1} \sum_{i=1}^{n} (\theta_i - \bar{\theta})^2
$$

$$
S_Y^2 = \frac{1}{n - 1} \sum_{i=1}^{n} (y_i - \bar{y})^2
$$

$$
S_{\theta Y} = \frac{1}{n - 1} \sum_{i=1}^{n} (\theta_i - \bar{\theta})(y_i - \bar{y})
$$

$$
S_{\theta Y}^2 = \left [\frac{1}{n - 1} \sum_{i=1}^{n} (\theta_i - \bar{\theta})(y_i - \bar{y})\right]^2
$$

Estos estimadores nos permiten aproximar la varianza muestral $S_Z^2$:


$$
S_Z^2 = \frac {(b-a)^2}{n}  \cdot \left (S_\theta^2 - \frac{S_{\theta Y}^2}{S_Y^2}\right)
$$

Esta es la **varianza muestral corregida**, que aproxima la mínima varianza posible al usar variables de control.

    <font size="10" color="lightblue" > Monte Carlo Crudo Variables Control </font>
</p>

```python
import numpy as np # Se importa la libreria numpy como "np"
def g(x): # Se define la funcion g(x) con parametro de entrada

# La funcion retorna el valor de la funcion 1/(x^2 + 1)

    return(1/((x**2) + 1))


def mcc_control(n, a, b): # se define la funcion aproximacion(N,a,b) con los parametros a, b, N
    X = []  # Esta lista guardara los numeros aleatorios X_1 ~ u(a,b)
    Y = []  # Esta lista guardara las evaluaciones en la funcion g(x)
    # Las lista deben estar afuera del ciclos for
    for i in range(n): # para i = 0 hasta i = n - 1
        x_i = np.random.uniform(a, b) # se genera un numero aleatorio que sigue una distribucion uniforme en [a,b]
        X.append(x_i) # se agrega el valor x_i a la lista X
        y_i = g(x_i) # se evalua el numero aleatorio en la funcion g(x)
        Y.append(y_i) # se agrega el valor y_i al arreglo Y
    Sth2 = np.var(X, ddof=1) # se calcula la varianza muestral del estimador θ con 1 grado de libertad
    Sy2 = np.var(Y, ddof=1) # se calcula la varianza muestral de la variable aleatoria Y con 1 grado de libertad
    Sthy = np.cov(X, Y, ddof=1)[0,1] # se calcula la covarianza muestral de las varaibles θ y Y con 1 grado de libertad
    Sz2 = Sth2 - (Sthy**2 / Sy2)

    Z = (b - a) * np.mean(Y) # estimacion de la integral I, usando el estimador Z pero Z = θ
    var_Z =  ((b - a)**2 / n) * (Sth2 - (Sthy**2) / Sy2) # calculo de la varianza muestral  de Z
    des_Z = np.sqrt(var_Z) # se calcula la desviacion estandar muestral de Z
    return Z, des_Z ,var_Z # el valor de salida es el valor del estimador Z, la varianza del estimador Z y la desviacion estandar del estimador Z


r_1 = mcc_control(10000, 0, 1) # se nombra la varaible "r_1" que guardara el valor que suelta la funcion "mcc_control(n,a,b) con
# n = 10000,a = 0 ,b = 1, que sera una tupla (Z, des_Z, var_Z )

print(
    f"\nResultado de la integral aproximada: {r_1[0]:.10f}" # mostrara el primer elemento del objeto r_1 con 10 cifras significativas, con variable de tipo flotante
    f"\nEl error estándar de aproximación (desviación estándar) es: {r_1[1]:.10f}"  # mostrara el segundo elemento de la tupla r_1 con 10 cifras significativas,
    # con variable de tipo flotante
    f"\nEl error estándar de aproximación al cuadrado (varianza) es: {r_1[2]:.10f}" # mostrara el tercer elemento de la tupla r_1 con 10 cifras significativas,
    # con variable de tipo flotante
)
```

```python
def mcc(n, a, b): # se define la funcion mcc(n,a,b) con los parametros n,a,b
    Y = []  # Esta lista guardara las muestras de g(x_i) para calcular la media
    # Las lista deben estar afuera del ciclos for
    for i in range(n): # para i = 0 hasta i = n - 1
        x_i = np.random.uniform(a, b) # se genera un numero aleatorio que sigue una distribucion uniforme en [a,b]
        y_i = g(x_i) # se evalua el numero aleatorio en la funcion g(x)
        Y.append(y_i) # se agrega el valor y_i al arreglo Y
    theta = (b - a) * np.mean(Y) # estimacion de la integral I, usando el estimador θ
    var_theta = (((b-a)**2)/n)* np.var(Y,ddof= 1) # se calcula la varianza muestral del estimador θ
    des_theta = np.sqrt(var_theta) # se calcula la desviacion muestral estandar del estimador θ
    return theta, des_theta ,var_theta # el valor de salida es el valor del estimador θ, la varianza muestral del estimador θ y la desviacion estandar muestral del estimador θ


r_2 = mcc(10000, 0, 1) # se nombra la varaible "r" que guardara el valor que suelta la funcion "mcc(n,a,b) con n = 10000,a = 0 ,b = 1,
# que sera una tupla (theta, var_theta, des_theta )

print(
    f"\nResultado de la integral aproximada: {r_2[0]:.10f}" # mostrara el primer elemento del objeto r_2 con 10 cifras significativas, con variable de tipo flotante
    f"\nEl error estándar de aproximación (desviación estándar) es: {r_2[1]:.10f}"  # mostrara el segundo elemento de la tupla r_2 con 10 cifras significativas,
    # con variable de tipo flotante
    f"\nEl error estándar de aproximación al cuadrado (varianza) es: {r_2[2]:.10f}" # mostrara el tercer elemento de la tupla r_2 con 10 cifras significativas,
    # con variable de tipo flotante
)
```

<p align="center">
    <font size="10" color="lightblue" >Metodo de Montecarlo Crudo vs Metodo de Monte Carlo Crudo con  Variables de Control  </font>
</p>

Se tiene que la varianza usando variables control  para estimar el valor $I$, se define como:
$$\mathrm{Var}(Z)$$

Se tiene que la varianza usando monte carlo crudo estandar para estimar el valor $I$, se define como:
$$\mathrm Var(\theta)$$


El porcentaje de reduccion de varianza ($\%_{\text{reduction}}$) del metodo de Monte Carlo Crudo respecto al metodo de Monte Carlo aplicando variables antiteticas se define como:

$$ \%_{\text{reduction}} = (1 - \frac {{Var}(Z)}{\mathrm Var(\theta)})\cdot 100\%$$

```python
def comp(a, b, n): # se define la funcion comp(a,b,n) con parametros a,b,n, esta funcion va a comparar la varianza #  del monte carlo crudo con
# la varianza del monte carlo usando variables de control

  var_control = r_1[2] # se nombra la variable var_control , que almacena el segundo parametro de la tupla r_1, que es la varianza usando el metodo de variables control
  var_crudo = r_2[2] # se nombra la variable "var_crudo", que almacena el segundo parametro de la tupla r_2, que es la varianza del monte carlo crudo normal
  c = np.abs(( 1 - (var_control/var_crudo) ) * 100) # usando la funcion np.abs (valor absoluto en numpy), calcula la comparacion absoluta del metodo de monte carlos
  # usando antiteticas y el metodo estandar
  return c # la funcion retorna el valor de la variable c

r_comp = comp(0, 1, 10000)
print(f"Hubo una reduccion del {r_comp:.5f}% en la varianza usando la tecnica de variables de control en Monte Carlo Crudo")
```
