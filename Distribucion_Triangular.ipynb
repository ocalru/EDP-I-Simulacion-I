```python
import numpy as np # Se importa la libreria numpy como "np"
import matplotlib.pyplot as plt # Se importa la libreria matplot.pyplot como "plt"
from scipy.stats import triang # De la libreria spicy.stats se importa la funcion "triang"
```

<p align="center">
    <font size="10" color="lightblue"> Definicion de la distribucion triangular. </font>
</p>



Sea $ a < c < b $ tres n√∫meros reales. Se define la **distribuci√≥n triangular** como una distribuci√≥n continua en el intervalo $ [a, b] $, que alcanza su valor m√°ximo en $ x = c $, y cuya densidad es lineal en los tramos $ [a, c] $ y $ [c, b] $.


Queremos construir una funci√≥n de densidad $ f(x) $ para la distribucion triangular, tal que:


- $ f(x) = 0 $ para $ x \notin [a, b] $
- $ f $ es continua en $ [a, b] $
- $ f $ es **linealmente creciente** en $ [a, c] $
- $ f $ es **linealmente decreciente** en $ [c, b] $
- $ \displaystyle \int_a^b f(x) \, dx = 1 $


Supongamos que $ f $ alcanza un valor m√°ximo $ h > 0 $ en $ x = c $, es decir
 $ f(c) = h $ , y queremos interpolar dos funciones lineales que pasen por los puntos:

- $ [ (a, 0) , (c, h) ] $  ‚ü∂ crecimiento lineal
- $ [ (c, h) , (b, 0) ] $  ‚ü∂ decrecimiento lineal



La pendiente entre dos puntos $ (x_1, y_1) $ y $ (x_2, y_2) $ es:

$$
m = \frac{y_2 - y_1}{x_2 - x_1}
$$


Para el intervalo $ [a, c] $, con los puntos $ (a, 0) $ y $ (c, h) $:

$$
m_1 = \frac{h - 0}{c - a} = \frac{h}{c - a}
$$

Para el intervalo $ [c, b] $, con los puntos $ (c, h) $ y $ (b, 0) $:

$$
m_2 = \frac{0 - h}{b - c} = -\frac{h}{b - c}
$$


Usando la forma punto-pendiente de una recta: $ y = m(x - x_1) + y_1
 $, se obtiene:

Para $ x \in [a, c] $:

$$
f(x) = m_1(x - a) = \frac{h(x - a)}{c - a}
\dots \, (1)
$$

Para $ x \in [c, b] $:

$$
f(x) = h + m_2(x - c) = \frac{h(b - x)}{b - c}
\dots \, (2)
$$



Para que $ f $ sea funci√≥n de densidad, debe cumplirse:

$$
\int_a^b f(x) \, dx = 1
$$

Partimos el dominio en dos integrales:

$$
\int_a^c f(x) \, dx + \int_c^b f(x) \, dx = 1
$$

Calculamos:

$$
\int_a^c \frac{h(x - a)}{c - a} dx = \frac{h}{c - a} \int_a^c (x - a) dx = \frac{h}{c - a} \cdot \frac{(c - a)^2}{2} = \frac{h(c - a)}{2}
$$

$$
\int_c^b \frac{h(b - x)}{b - c} dx = \frac{h}{b - c} \int_c^b (b - x) dx = \frac{h}{b - c} \cdot \frac{(b - c)^2}{2} = \frac{h(b - c)}{2}
$$

Entonces:

$$
\int_a^b f(x) dx = \frac{h(c - a)}{2} + \frac{h(b - c)}{2} = \frac{h(b - a)}{2}
$$

Imponiendo la condici√≥n de normalizaci√≥n:

$$
\frac{h(b - a)}{2} = 1 \quad \Rightarrow \quad h = \frac{2}{b - a}
$$

Sustituyendo $ h = \dfrac{2}{b - a} $ en las ecuaciones $(1)$ y $(2)$:


Para $ x \in [a, c] $:

$$
f(x) = \frac{2}{b - a} \cdot \frac{x - a}{c - a}
$$

Para $ x \in [c, b] $:

$$
f(x) = \frac{2}{b - a} \cdot \frac{b - x}{b - c}
$$




Se tiene, que la funci√≥n de densidad resultante es:

$$
f(x) =
\begin{cases}
\displaystyle \frac{2(x - a)}{(b - a)(c - a)}, & \text{si } a \leq x \leq c \\
\displaystyle \frac{2(b - x)}{(b - a)(b - c)}, & \text{si } c < x \leq b \\
0, & \text{en otro caso}
\end{cases}
$$






<p align="center">
    <font size="10" color="lightblue"> Calculo de la funci√≥n de Distribuci√≥n Acumulada (CDF) de una Distribuci√≥n Triangular </font>
</p>



La funci√≥n de distribuci√≥n acumulada $ F(x) $ es  la **probabilidad de que la variable aleatoria $ X $ sea menor o igual a $ x $**:

$$
F(x) = P(X \leq x) = \int_{-\infty}^{x} f(t) \, dt
$$


Sea una variable aleatoria continua con distribuci√≥n **triangular** en el intervalo $ [a, b] $, con valor maximo $ c $, donde $ a < c < b $. La funci√≥n de densidad es:

$$
f(x) =
\begin{cases}
0, & x < a \\
\frac{2(x - a)}{(b - a)(c - a)}, & a \le x < c \\
\frac{2(b - x)}{(b - a)(b - c)}, & c < x \le b \\
0, & x > b
\end{cases}
$$
$ U = F(X) \sim \text{Uniform}(0,1) $


Si $ X $ es una variable aleatoria continua con funci√≥n de distribuci√≥n acumulada (CDF) $ F_X(x) $, y $ F_X $ es estrictamente creciente y continua, entonces:

Y viceversa: si $ U \sim \text{Uniform}(0,1) $, y definimos

$$
X = F_X^{-1}(U),
$$

entonces:

$$
X \sim \text{Distribuci√≥n con CDF } F_X.
$$


La funci√≥n de distribuci√≥n acumulada $ F(x) $ nos da la probabilidad de que $ X \leq x $. Es decir:

$$
F(x) = \mathbb{P}(X \leq x)
$$

Entonces, cuando definimos \( U = F(X) \), estamos diciendo que \( U \) representa la **probabilidad acumulada** de un valor aleatorio de \( X \).  
Como \( X \) toma valores seg√∫n su distribuci√≥n, \( F(X) \) recorre el intervalo \( (0,1) \) de manera uniforme, **siempre que \( F \) sea estrictamente creciente y continua**.

---

### üìê Demostraci√≥n (esbozo formal)

Queremos demostrar que si \( X \) tiene CDF \( F_X(x) \), entonces \( U = F_X(X) \sim \text{Uniform}(0,1) \).

**Paso 1:** Hallar la CDF de \( U = F_X(X) \)

Sea \( u \in (0,1) \). Entonces:

\[
\mathbb{P}(U \leq u) = \mathbb{P}(F_X(X) \leq u)
\]

Como \( F_X \) es estrictamente creciente, aplicamos su inversa:

\[
\mathbb{P}(F_X(X) \leq u) = \mathbb{P}(X \leq F_X^{-1}(u)) = F_X(F_X^{-1}(u)) = u
\]

Por lo tanto, la CDF de \( U \) es:

\[
F_U(u) = \mathbb{P}(U \leq u) = u
\]

Esto es exactamente la CDF de una distribuci√≥n uniforme en \( (0,1) \).

---

### üîÅ Inversa:

Si \( U \sim \text{Uniform}(0,1) \), entonces:

\[
X = F^{-1}(U) \sim F
\]

Esto es la **transformada inversa**, utilizada para generar variables aleatorias con una distribuci√≥n deseada a partir de una variable uniforme.

---

## ‚úÖ Correspondencia entre CDF y Uniforme: Tu observaci√≥n

T√∫ observaste que:

- La CDF de cualquier variable continua \( X \), si es estrictamente creciente y continua, transforma valores de \( X \) al intervalo \( (0,1) \).

- La CDF de \( U \sim \text{Uniform}(0,1) \) tambi√©n es creciente y continua.

Entonces:

\[
F_X(X) \sim \text{Uniform}(0,1)
\]

Esta es exactamente la idea del teorema: **toda CDF estrictamente creciente y continua transforma su variable en una variable uniforme**.

---

## üß† En resumen:

‚úÖ Tu razonamiento es correcto:

- Tanto \( F_X \) como la CDF de una variable uniforme son funciones **crecientes** y **continuas**.

- Entonces, aplicar \( F_X \) a \( X \sim F \) produce una variable con las mismas propiedades que una uniforme en \( (0,1) \).

\[
\boxed{F_X(X) \sim \text{Uniform}(0,1)}
\]

Y de manera inversa:

\[
\boxed{X = F_X^{-1}(U) \Rightarrow X \sim F}
\]



La funci√≥n de distribuci√≥n acumulada $ F(x) $ se obtiene integrando la funci√≥n de densidad $ f(x) $ en cada tramo definido.

Caso 1 ($ x < a $):

$$
F(x) = 0
$$


Caso 2 ($ a \leq x \leq c $):

En este caso, la funci√≥n de distribuci√≥n acumulada (CDF) se calcula integrando la funci√≥n de densidad $ f(t) $ en el intervalo $ [a, x] $, que est√° contenido en el intervalo $ [a, c] $.

Se tiene la integral:

$$
\int_a^x f(t) \, dt = \int_a^x \frac{2(t - a)}{(b - a)(c - a)} \, dt
$$

Sacando las constantes:

$$
= \frac{2}{(b - a)(c - a)} \int_a^x (t - a) \, dt
$$

$$
= \frac{2}{(b - a)(c - a)} \left[ \frac{(t - a)^2}{2} \right]_a^x
$$

Evaluamos los l√≠mites:

$$
= \frac{2}{(b - a)(c - a)} \left( \frac{(x - a)^2}{2} - 0 \right)
$$

$$
= \frac{(x - a)^2}{(b - a)(c - a)}
$$

La CDF en este caso es:

$$
F(x) = \frac{(x - a)^2}{(b - a)(c - a)}
$$


Caso 3 ($ c < x \leq b $):

Para calcular la funci√≥n de distribuci√≥n acumulada (CDF) cuando $ c < x \leq b$ , se divide la integral en dos partes y se suman ambas integrales.

- La primera integral corresponde a la densidad en el intervalo $ [a, c] $, donde la funcion de denisdad $ f(x) $ es **linealmente creciente**.
- La segunda integral corresponde a la densidad en el intervalo $ [c, x] $, donde la funcion $ f(x) $ **linealmente decreciente**.

Se tiene que:

$$
F(x) = \int_a^c f(t) \, dt + \int_c^x f(t) \, dt
$$

Calculando la integral en $ [a, c] $:

$$
\int_a^c \frac{2(t - a)}{(b - a)(c - a)} \, dt
$$

Sacando las constantes:

$$
= \frac{2}{(b - a)(c - a)} \int_a^c (t - a) \, dt
$$


$$
= \frac{2}{(b - a)(c - a)} \left[ \frac{(t - a)^2}{2} \right]_a^c
$$

Evaluando la integral definida:

$$
= \frac{2}{(b - a)(c - a)} \left( \frac{(c - a)^2}{2} \right)
$$

$$
= \frac{(c - a)}{(b - a)}
$$

Calculando la integral en $ [c, x] $:


$$
\int_c^x \frac{2(b - t)}{(b - a)(b - c)} \, dt
$$

Sacando las constantes:

$$
= \frac{2}{(b - a)(b - c)} \int_c^x (b - t) \, dt
$$


$$
= \frac{2}{(b - a)(b - c)} \left[ \frac{(b - t)^2}{2} \right]_c^x
$$

Evaluando la integral definida:

$$
= \frac{2}{(b - a)(b - c)} \left( \frac{(b - c)^2}{2} - \frac{(b - x)^2}{2} \right)
$$

$$
= \frac{(b - c)^2 - (b - x)^2}{(b - a)(b - c)}
$$

Finalmente, sumando ambas √°reas:

$$
F(x) = \frac{(c - a)}{(b - a)} + \frac{(b - c)^2 - (b - x)^2}{(b - a)(b - c)}
$$

Simplificando:

$$
F(x) = 1 - \frac{(b - x)^2}{(b - a)(b - c)}
$$


Caso 4 ($ x > b $):


$$
F(x) = 1
$$

Entonces la funcion de distribucion acumulada $ F(x) $, queda definida como:


$$
F(x) =
\begin{cases}
0 & \text{si } x < a \\
\frac{(x - a)^2}{(b - a)(c - a)} & \text{si } a \leq x \leq c \\
1 - \frac{(b - x)^2}{(b - a)(b - c)} & \text{si } c < x \leq b \\
1 & \text{si } x > b \\
\end{cases}
$$


En el punto $ x = c $, se alcanza el **m√°ximo de la funci√≥n de densidad** de la distribuci√≥n triangular.  

üîπ Para $ x = c^- $ (a la izquierda de $c$):

$$
F(c^-) = \frac{(c - a)^2}{(b - a)(c - a)} = \frac{c - a}{b - a}
$$

üîπ Para $ x = c^+ $ (a la dereche del punto  $c$ ):

$$
F(c^+) = 1 - \frac{(b - c)^2}{(b - a)(b - c)} = 1 - \frac{b - c}{b - a} = \frac{c - a}{b - a}
$$

Ambas evaluaciones coinciden:  

$$
F(c^-) = F(c^+) = \frac{c - a}{b - a}
$$


$$
F(c^-) = F(c^+) = \frac{c - a}{b - a}
$$

Esto confirma que $ F(x) $ es continua en $ x = c $, como debe ser toda funci√≥n de distribuci√≥n acumulada. Al ser continua, entonces se tiene que:

$$
F(c) = F(c^-) = F(c^+) = \frac{c - a}{b - a}
$$



Como $ U = F(X) $ en $ U \in [0, 1] $, entonces **cuando invertimos $ F $**, debemos separar $ U $ en dos tramos:

- Si $ 0 \leq U < \frac{c - a}{b - a} $ ‚Üí corresponde al tramo $ a \leq X < c $
- Si $ \frac{c - a}{b - a} \leq U \leq 1 $ ‚Üí corresponde al tramo $ c \leq X \leq b $


La expresi√≥n completa para obtener $ X $ a partir de $ U \sim \text{Uniform}(0, 1) $ es:

$$
X =
\begin{cases}
a + \sqrt{U \cdot (b - a)(c - a)} & \text{si } 0 \leq U < \frac{c - a}{b - a} \\
b - \sqrt{(1 - U) \cdot (b - a)(b - c)} & \text{si } \frac{c - a}{b - a} \leq U \leq 1
\end{cases}
$$








```python


# Par√°metros de la distribuci√≥n triangular

a = 1
b = 3
c = 5

# N√∫mero de muestras
n = 100000

# M√©todo de la transformada inversa
def triangular_inverse_transform(u, a, b, c):
    Fc = (b - a) / (c - a)
    x = np.where(
        u < Fc,
        a + np.sqrt((b - a) * (c - a) * u),
        c - np.sqrt((c - b) * (c - a) * (1 - u))
    )
    return x

# Generar muestras
u = np.random.uniform(0, 1, n)
samples = triangular_inverse_transform(u, a, b, c)

# Se calcula la media de la muestra
media_empirica = np.mean(samples)
desv_est_empirica = np.std(samples)

# C√°lculos te√≥ricos
media_teorica = (a + b + c) / 3
var_teorica = (a**2 + b**2 + c**2 - a*b - a*c - b*c) / 18
desv_est_teorica = np.sqrt(var_teorica)

# Mostrar resultados
print(f"Media emp√≠rica: {media_empirica:.4f}")
print(f"Desviaci√≥n est√°ndar emp√≠rica: {desv_est_empirica:.4f}")
print(f"Media te√≥rica: {media_teorica:.4f}")
print(f"Desviaci√≥n est√°ndar te√≥rica: {desv_est_teorica:.4f}")

# Graficar histograma
plt.figure(figsize=(10, 6))
plt.hist(samples, bins=100, density=True, alpha=0.6, color='skyblue', label='Histograma (emp√≠rico)')

# Graficar distribuci√≥n te√≥rica
x = np.linspace(a, c, 1000)
pdf_teorica = triang.pdf(x, (b - a) / (c - a), loc=a, scale=c - a)
plt.plot(x, pdf_teorica, 'r-', lw=2, label='PDF te√≥rica')

plt.title("Distribuci√≥n Triangular (a=1, b=3, c=5)")
plt.xlabel("Valor")
plt.ylabel("Densidad")
plt.legend()
plt.grid(True)
plt.show()
```
