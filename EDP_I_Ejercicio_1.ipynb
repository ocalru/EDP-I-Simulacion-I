Grafica de la funcion: <font color='red'>$y=x^2$ <font color='black'> en el intervalo $ [-10, 10]$ |



images.jpg

```python
import numpy as np
import matplotlib.pyplot as plt

# Definir el rango de valores para x
x = np.linspace(-10, 10, 400)  # 400 puntos entre -10 y 10
y = x**2  # Funcin cuadrtica

# Crear la grfica
plt.plot(x, y, label=r'$y = x^2$', color='b')

# Etiquetas y ttulo
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grfica de y = x^2')

# Mostrar la cuadrcula
plt.grid(True, linestyle='--', alpha=0.7)

# Mostrar la leyenda
plt.legend()

# Mostrar la grfica
plt.show
```
