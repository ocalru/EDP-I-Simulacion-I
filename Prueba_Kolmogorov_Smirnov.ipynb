El estadístico de prueba de Kolmogorov se define como:

$$
D_n = \sup_{x} \left| F_n(x) - F(x) \right|
$$

donde:

•$F_n(x) = \frac{i}{n}$ es la función de distribución acumulada empírica basada en la muestra $( x_1, x_2, \dots, x_n )$, con $ k $ el número de observaciones menores o iguales a $ x $.

• $ F(x) $ es la función de distribución acumulada teórica (como por ejemplo la distribucion exponencial, uniforme, etc)

• El estadístico $D_n $ mide la máxima diferencia absoluta entre la distribución empírica y la distribución teórica.

Para calcular $D_n$, seguimos estos pasos:

1. Ordenamos la muestra $( x_1, x_2, \dots, x_n )$ tal que $( x_1 \leq x_2 \leq \dots \leq x_n )$

2. Calculamos las estadísticas $( D_- )$ y $( D_+ )$

   $$
   D_- = \max_{1 \leq i \leq n} \left( \frac{i}{n} - F(x_i) \right)
   $$

   $$
   D_+ = \max_{1 \leq i \leq n} \left( F(x_i) - \frac{i-1}{n} \right)
   $$

3. El estadístico de prueba es:

   $$
   D_n = \max(D_-, D_+)
   $$

Se tiene que $F(x)$ es una distribucion uniforme, su funcion acumulada se define como:

$$
F(x) = P(X \leq x) =
\begin{cases}
0, & x < a \\
\frac{x - a}{b - a}, & a \leq x \leq b \\
1, & x > b
\end{cases}
$$

Para el caso específico de $ U(0,1) $, donde $ a = 0 $ y $ b = 1 $, la funcion de distribucion acumulada se reduce a:

$$
F(x) =
\begin{cases}
0, & x < 0 \\
\frac{x - 0}{1 - 0} = x, & 0 \leq x \leq 1 \\
1, & x > 1
\end{cases}
$$

```python
# Nivel de significancia (α)
alpha = 0.05

# Hipótesis nula y alternativa:

# H0: Los datos siguen una distribución uniforme U(0,1)
# H1: Los datos no siguen una distribución uniforme U(0,1)

# Realizamos la prueba de hipótesis
if p_value < alpha:
    conclusion = "Rechazamos la hipótesis nula. Los datos NO siguen una distribución uniforme U(0,1)."
else:
    conclusion = "No rechazamos la hipótesis nula. Los datos SIGUEN una distribución uniforme U(0,1)."

# Mostrar conclusión
print(conclusion)
```
