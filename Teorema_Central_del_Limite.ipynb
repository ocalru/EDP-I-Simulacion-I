    <font size="10" color="lightblue"> Aplicación Teorema Central del Límite (TCL) al experimento de extracción de cartas.  </font>
</p>





Sea $ \Omega $ el espacio muestral correspondiente al conjunto de todas las **cartas posibles** en un mazo estándar de 52 cartas, se realiza una extraccion de una carta del mazo, en cada extraccion se regresa la carta seleccionada.

Definimos el espacio muestral de la extraccion de una carta:


$$
\Omega := \{C_1, C_2, \dots, C_{52}\}
$$

Cada $ C_j $ representa una carta específica del mazo estándar. Por ejemplo:

- $ C_1 = \text{as de tréboles} $
- $ C_2 = \text{2 de tréboles} $
- $ \vdots $
- $ C_{13} = \text{rey de tréboles} $
- $ C_{14} = \text{as de corazones} $
- $ \vdots $
- $ C_{26} = \text{rey de corazones} $
- $ C_{27} = \text{as de diamantes} $
- $ \vdots $
- $ C_{39} = \text{rey de diamantes} $
- $ C_{40} = \text{as de picas} $
- $ \vdots $
- $ C_{52} = \text{rey de picas} $

Cada elemento de $ \Omega $ representa una carta individual del mazo.


Se define la variable aleatoria $ X_i $ como el valor numérico de la carta extraída en el $ i-ésimo$ ensayo. Los valores numéricos de las cartas se asignan de la siguiente manera:

- $ A = 1 $
- $ J = 11 $
- $ Q = 12 $
- $ K = 13 $

y los valores de las cartas numéricas $ 2, 3, \dots, 10 $ corresponden a sus respectivos valores numéricos. Entonces, el conjunto de posibles valores para  $ X_i $ es:

$$
\mathcal{X} := \{1, 2, \dots, 13\}
$$



Definimos una **variable aleatoria discreta**:

$$
X_i: \Omega \longrightarrow \mathcal{X}
$$

A cada elemento  del espacio muestral $ \Omega $ se le asigna un valor en el conjunto $ \mathcal{X} $, el cual representa el valor numérico de la carta extraída en el ensayo número $ i $.

Sea $ \{X_1, X_2, \dots, X_n\} $ una secuencia de variables aleatorias **i.i.d.** (idénticamente distribuidas e independientes), donde cada $ X_i $ representa el valor numérico de la carta extraída en el ensayo $ i $, bajo **extracción con reemplazo**.


Para todo $ k \in \mathcal{X} $, se cumple:

$$
P(X_i = k) = \frac{4}{52} = \frac{1}{13}, \quad \text{ya que hay 4 cartas por cada valor}
$$



El valor esperado de cada $ X_i $ es:

$$
\mathbb{E}[X_i] = \sum_{k=1}^{13} k \cdot \frac{1}{13} = \frac{1}{13} \sum_{k=1}^{13} k = \frac{1}{13} \cdot \frac{13 \cdot 14}{2} = 7
$$

El valor esperado al cuadrado de cada $ X_i $ es:

$$
\mathbb{E}[X_i^2] = \sum_{k=1}^{13} k^2 \cdot \frac{1}{13} = \frac{1}{13} \sum_{k=1}^{13} k^2 = \frac{1}{13} \cdot \frac{13 \cdot 14 \cdot 27}{6} = 63
$$

La varianza de cada $ X_i $ es:

$$
\text{Var}(X_i) = \mathbb{E}[X_i^2] - (\mathbb{E}[X_i])^2 = 63 - 7^2 = 63 - 49 = 14
$$


Sea $ \{X_1, X_2, \dots, X_n\} $ una sucesión de variables aleatorias **i.i.d.** con esperanza $ \mu = \mathbb{E}[X_i] $ y varianza finita $ \sigma^2 = \text{Var}(X_i) $. Entonces, el TCL establece que:

$$
Z_n := \frac{\overline{X}_n - \mu}{\sigma / \sqrt{n}} \xrightarrow{d} \mathcal{N}(0,1) \quad \text{cuando } n \to \infty
$$

Donde:

- $ \overline{X}_n = \frac{1}{n} \sum_{i=1}^{n} X_i $ es la **media muestral**
- $ \mu = \mathbb{E}[X_i] $ es la **media poblacional**
- $ \sigma^2 = \operatorname{Var}(X_i) $ es la **varianza poblacional**
- $ n $ es el **tamaño de la muestra**
- La notación $ \xrightarrow{d} \mathcal{N}(0, 1) $ significa **convergencia en distribución** hacia una **distribución normal estándar** (media 0 y varianza 1).

Esto quiere decir que, aunque los $ X_i $ no sigan una distribución normal, cuando el tamaño de la muestra $ n $ es suficientemente grande, la distribución de la media muestral estandarizada $ Z_n $ se aproxima a una distribución normal estándar.

Aplicando el TLC con:


- $ \mu = 7 $
- $ \sigma^2 = 14 $
- $ \sigma = \sqrt{14} $

Entonces:

$$
\overline{X}_n \approx \mathcal{N}\left( 7, \frac{14}{n} \right)
$$

Por ejemplo, si $ n = 12 $ ( $ 12 $ extracciones )

$$
\overline{X}_{12} \sim \mathcal{N}\left( 7, \frac{14}{12} \right) = \mathcal{N}(7, 1.167)
$$




Tras 12 extracciones se obtuviero los siguientes valores:

$$
x = \{6, 9, 5, 13, 4, 7, 8, 5, 6, 10, 6, 7\}
$$


$$
\overline{x}_{12} = \frac{1}{12} \sum_{i=1}^{12} x_i = \frac{86}{12} \approx 7.17
$$


$$
Z = \frac{\overline{x}_{12} - \mu}{\sigma / \sqrt{n}} = \frac{7.17 - 7}{\sqrt{14/12}} \approx \frac{0.17}{1.08} \approx 0.157
$$

La probabilidad de que $ Z \leq 0.157 $ se calcula como:

$$
P(Z \leq 0.157) = \int_{-\infty}^{0.157} \frac{1}{\sqrt{2\pi}} e^{-z^2 / 2} \, dz \approx 0.562
$$

Donde $ Z $ sigue una distribución normal estándar $ \mathcal{N}(0, 1) $. Esta integral es la función de distribución acumulada (CDF) de la normal estándar.
